对应练习文件: `exer0501.ipynb` ~ `exer0504.ipynb`

## 假设检验基本概念

假设检验中，**原假设** $H_0$ 与**备择假设** $H_1$ 是一对互斥的陈述。若样本观测值落入**拒绝域** $D$，则拒绝 $H_0$。检验可能犯两类错误：**第一类错误（拒真）** $\alpha=P\{\text{拒绝}H_0|H_0\text{为真}\}$，**第二类错误（纳伪）** $\beta=P\{\text{接受}H_0|H_1\text{为真}\}$。**功效函数**定义为 $\varphi(\theta)=P\{\text{拒绝}H_0|\theta\}$，**显著性水平**即犯第一类错误的最大概率。

### 正态总体均值检验

设 $H_0:\mu=0$ vs $H_1:\mu=1$，拒绝域 $D=\{\overline{X}\ge0.98\}$。当 $H_0$ 真时 $\alpha=1-\Phi(\sqrt{n}(0.98-\mu_0))$，当 $H_1$ 真时 $\beta=\Phi(\sqrt{n}(0.98-\mu_1))$。给定 $\alpha$ 可反求所需样本容量 $n$。

### 一般分布的检验

对于 $H_0: f(x)=1$（均匀分布）vs $H_1: f(x)=2x$（三角形分布），拒绝域 $D=\{X>2/3\}$，可算得 $\alpha=P\{X>2/3|H_0\}$，$\beta=P\{X\le2/3|H_1\}$。

### 均匀分布的顺序统计量检验

对于 $H_0:\theta=1$ vs $H_1:\theta>1$，$X\sim U[0,\theta]$，拒绝域 $D=\{X_{(10)}\ge0.95\}$。当 $\theta=1$ 时 $\alpha=1-0.95^{10}$。功效函数为
$$\varphi(\theta)=\begin{cases}
0, & \theta<0.95, \\
1-(0.95/\theta)^{10}, & \theta\ge0.95.
\end{cases}$$

## 正态总体的假设检验

### $\sigma$ 已知时 $\mu$ 的检验（$Z$ 检验）

检验统计量为 $Z=\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}$，当 $H_0$ 成立时 $Z\sim N(0,1)$。双边检验 $H_0:\mu=\mu_0$ vs $H_1:\mu\ne\mu_0$ 的拒绝域为 $\{|Z|\ge u_{\alpha/2}\}$；单边检验 $H_0:\mu\ge\mu_0$ vs $H_1:\mu<\mu_0$ 的拒绝域为 $\{Z\le -u_{\alpha}\}$。第二类错误概率为
$$\beta=\Phi\left(u_{\alpha/2}+\frac{\mu_0-\mu_1}{\sigma/\sqrt{n}}\right)-\Phi\left(-u_{\alpha/2}+\frac{\mu_0-\mu_1}{\sigma/\sqrt{n}}\right).$$

### $\sigma$ 未知时 $\mu$ 的检验（$t$ 检验）

检验统计量为 $T=\frac{\overline{X}-\mu_0}{S/\sqrt{n}}$，当 $H_0$ 成立时 $T\sim t(n-1)$。单边拒绝域为 $\{T>t_\alpha(n-1)\}$，双边为 $\{|T|\ge t_{\alpha/2}(n-1)\}$。

### $\sigma^2$ 的检验（$\chi^2$ 检验）

检验统计量为 $\chi^2=\frac{(n-1)S^2}{\sigma_0^2}$，当 $H_0$ 成立时 $\chi^2\sim\chi^2(n-1)$。双边检验 $H_0:\sigma=\sigma_0$ vs $H_1:\sigma\ne\sigma_0$ 的拒绝域为 $\{\chi^2\ge\chi^2_{\alpha/2}(n-1)\text{ 或 }\chi^2\le\chi^2_{1-\alpha/2}(n-1)\}$。单边检验 $H_0:\sigma^2\ge\sigma_0^2$ vs $H_1:\sigma^2<\sigma_0^2$ 的拒绝域为 $\{\chi^2\le\chi^2_{1-\alpha}(n-1)\}$。

### 两正态总体的假设检验

对于 $H_0:\mu_1\le\mu_2$ vs $H_1:\mu_1>\mu_2$，若 $\sigma_1^2,\sigma_2^2$ 已知，检验统计量为 $Z=\frac{\overline{X}-\overline{Y}}{\sqrt{\sigma_1^2/n_1+\sigma_2^2/n_2}}\sim N(0,1)$；若 $\sigma_1^2=\sigma_2^2$ 未知，检验统计量为 $T=\frac{\overline{X}-\overline{Y}}{S_w\sqrt{1/n_1+1/n_2}}\sim t(n_1+n_2-2)$，其中 $S_w^2=\frac{(n_1-1)S_1^2+(n_2-1)S_2^2}{n_1+n_2-2}$。

对于 $H_0:\sigma_1^2=\sigma_2^2$ 的检验，检验统计量为 $F=\frac{S_1^2}{S_2^2}\sim F(n_1-1,n_2-1)$（当 $H_0$ 成立），拒绝域为 $\{F\ge F_{\alpha/2}(n_1-1,n_2-1)\text{ 或 }F\le F_{1-\alpha/2}(n_1-1,n_2-1)\}$。

## 非正态总体的假设检验

### 指数分布总体的检验

对于 $X\sim E(\lambda)$，检验 $H_0:\mu\ge1000$ vs $H_1:\mu<1000$ 时，检验统计量为 $Q=2n\lambda\overline{X}\sim\chi^2(2n)$，拒绝域为 $\{Q\le\chi^2_{1-\alpha}(2n)\}$。

比较两指数总体 $H_0:\lambda_1=\lambda_2$ vs $H_1:\lambda_1\ne\lambda_2$ 时，利用 $2m\lambda_1\overline{X}\sim\chi^2(2m)$ 和 $2n\lambda_2\overline{Y}\sim\chi^2(2n)$，可得 $F=\frac{\overline{X}}{\overline{Y}}\sim F(2m,2n)$。

### Poisson分布总体的检验

对于 $X\sim P(\lambda)$，大样本下检验 $H_0:\lambda=\lambda_0$ vs $H_1:\lambda\ne\lambda_0$，由中心极限定理有 $Z=\frac{\overline{X}-\lambda_0}{\sqrt{\lambda_0/n}}\sim N(0,1)$，拒绝域为 $\{|Z|\ge u_{\alpha/2}\}$。

### 比例的检验

对于 $X\sim B(n,p)$，样本比例 $\widehat{p}=X/n$。检验 $H_0:p=0.5$ vs $H_1:p\ne0.5$ 时，$Z=\frac{\widehat{p}-p_0}{\sqrt{p_0(1-p_0)/n}}\sim N(0,1)$，拒绝域为 $\{|Z|\ge u_{\alpha/2}\}$。比例检验的本质是二项分布的正态近似，大样本条件为 $np\ge5$ 且 $n(1-p)\ge5$。

## 区间估计与假设检验的对偶性

置信水平为 $1-\alpha$ 的置信区间与显著性水平为 $\alpha$ 的假设检验之间存在对偶关系。若 $[\hat{\theta}_L,\hat{\theta}_U]$ 是 $\theta$ 的置信水平 $1-\alpha$ 的置信区间，则检验 $H_0:\theta=\theta_0$ 的拒绝域为 $\{\theta_0\notin[\hat{\theta}_L,\hat{\theta}_U]\}$。

### 利用对偶性导出置信区间

对于单个正态总体的 $\sigma^2$，检验统计量 $Q=\frac{(n-1)S^2}{\sigma^2}\sim\chi^2(n-1)$，利用对偶性得置信区间
$$\left[\frac{(n-1)S^2}{\chi^2_{1-\alpha/2}(n-1)},\;\frac{(n-1)S^2}{\chi^2_{\alpha/2}(n-1)}\right].$$

对于两正态总体的 $\sigma_1^2/\sigma_2^2$，检验统计量 $Q=\frac{S_1^2/\sigma_1^2}{S_2^2/\sigma_2^2}\sim F(m-1,n-1)$，得置信区间
$$\left[\frac{S_1^2}{S_2^2}F_{\alpha/2}(n-1,m-1),\;\frac{S_1^2}{S_2^2}F_{1-\alpha/2}(n-1,m-1)\right].$$

对偶性是数理统计的核心思想之一：区间估计和假设检验是同一问题的两种表述，已知检验方法可构造置信区间，反之亦然。
