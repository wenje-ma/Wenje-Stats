# 第 5 章 假设检验


> **本章导读**: 参数估计是参数是多少? 假设检验是参数是否等于某个值? 本章从两类错误和功效函数出发, 介绍正态总体的 z 检验、t 检验、chi^2 检验, 以及其他分布的检验方法.


---


## 5.1 假设检验的基本概念


原假设 H0 vs 备择假设 H1


第一类错误(弃真): alpha = P{拒绝 H0 | H0 为真}


第二类错误(纳伪): beta = P{接受 H0 | H1 为真}


```python
# 代码演示:两类错误的模拟
import numpy as np
np.random.seed(42)

# H0: mu=0, H1: mu=1, sigma=1, n=10
# 拒绝域: X_bar > c
n=10
N_sim=50000
alpha=0.05
from scipy.stats import norm
c=norm.ppf(1-alpha)/np.sqrt(n)  # 拒绝域临界值

# 模拟第一类错误(H0 为真时拒绝)
samples_h0=np.random.normal(0,1,(N_sim,n))
xbar_h0=np.mean(samples_h0,axis=1)
type1=np.mean(xbar_h0>c)

# 模拟第二类错误(H1 为真时接受)
samples_h1=np.random.normal(1,1,(N_sim,n))
xbar_h1=np.mean(samples_h1,axis=1)
type2=np.mean(xbar_h1<=c)

print('模拟第一类错误率:', round(type1,4), '(设定 alpha=0.05)')
print('模拟第二类错误率:', round(type2,4))
print('功效:', round(1-type2,4))
```

## 5.2 正态总体的假设检验


### 5.2.1 z 检验(sigma 已知时 mu 的检验)


检验统计量: Z = (X_bar-mu0)/(sigma/sqrt(n)) ~ N(0,1)


```python
# 代码演示:z 检验
import numpy as np
from scipy.stats import norm
np.random.seed(42)

# H0: mu=26, H1: mu!=26, sigma=5.2, n=16
mu0,sigma,n=26,5.2,16
sample=np.random.normal(27.56,sigma,n)
xbar=np.mean(sample)
z=(xbar-mu0)/(sigma/np.sqrt(n))
p_value=2*(1-norm.cdf(abs(z)))
print('z 统计量:', round(z,4))
print('p 值:', round(p_value,4))
if p_value<0.05: print('拒绝 H0')
else: print('无法拒绝 H0')
```

### 5.2.2 t 检验(sigma 未知时 mu 的检验)


检验统计量: T = (X_bar-mu0)/(S/sqrt(n)) ~ t(n-1)


```python
# 代码演示:t 检验
import numpy as np
from scipy.stats import t as t_dist
np.random.seed(42)
mu0,n=26,16
sample=np.random.normal(27.56,5.2,n)
xbar=np.mean(sample)
s=np.std(sample,ddof=1)
t_stat=(xbar-mu0)/(s/np.sqrt(n))
p_value=2*(1-t_dist.cdf(abs(t_stat),n-1))
print('t 统计量:', round(t_stat,4))
print('p 值:', round(p_value,4))
if p_value<0.05: print('拒绝 H0')
else: print('无法拒绝 H0')
```

### 5.2.3 chi^2 检验(sigma 的检验)


检验统计量: Q = (n-1)S^2/sigma0^2 ~ chi^2(n-1)


```python
# 代码演示:chi^2 检验
import numpy as np
from scipy.stats import chi2
np.random.seed(42)
sigma0,n=0.048,5
sample=np.random.normal(0,0.07,n)
s2=np.var(sample,ddof=1)
q=(n-1)*s2/sigma0**2
p_value=2*min(chi2.cdf(q,n-1),1-chi2.cdf(q,n-1))
print('Q 统计量:', round(q,4))
print('p 值:', round(p_value,4))
if p_value<0.05: print('拒绝 H0')
else: print('无法拒绝 H0')
```

## 5.3 其他分布的假设检验


### 5.3.1 指数分布检验


检验统计量: Q=2n*lambda0*X_bar ~ chi^2(2n)


```python
# 代码演示:指数分布检验
import numpy as np
from scipy.stats import chi2
np.random.seed(42)
mu0,n=1000,4
sample=np.random.exponential(mu0,n)
xbar=np.mean(sample)
q=2*n*xbar/mu0
p=chi2.cdf(q,2*n)
print('Q 统计量:', round(q,4))
print('p 值:', round(p,4))
if p<0.05: print('拒绝 H0')
else: print('无法拒绝 H0')
```

### 5.3.2 比例 z 检验


```python
# 代码演示:比例 z 检验
import numpy as np
from scipy.stats import norm
np.random.seed(42)
p0,n=0.5,1000
sample=np.random.binomial(1,0.45,n)
p_hat=np.mean(sample)
z=(p_hat-p0)/np.sqrt(p0*(1-p0)/n)
p_value=2*(1-norm.cdf(abs(z)))
print('样本比例:', round(p_hat,4))
print('z 统计量:', round(z,4))
print('p 值:', round(p_value,4))
if p_value<0.05: print('拒绝 H0')
else: print('无法拒绝 H0')
```

## 5.4 置信区间与假设检验的对偶性


参数的置信水平为 1-alpha 的置信区间, 恰好是显著性水平为 alpha 的检验的接受域.


```python
# 代码演示:对偶性验证
import numpy as np
from scipy.stats import norm,t as t_dist
np.random.seed(42)
true_mu,true_sigma=5,2
n=25
alpha=0.05
sample=np.random.normal(true_mu,true_sigma,n)
xbar=np.mean(sample)
s=np.std(sample,ddof=1)
tv=t_dist.ppf(1-alpha/2,n-1)
ci=[xbar-tv*s/np.sqrt(n), xbar+tv*s/np.sqrt(n)]
print('mu 的 95% 置信区间:', [round(x,4) for x in ci])

# 对偶性:mu0 在区间内 <=> 不拒绝 H0: mu=mu0
for mu0 in [4,4.5,5,5.5,6]:
    t_stat=(xbar-mu0)/(s/np.sqrt(n))
    p_val=2*(1-t_dist.cdf(abs(t_stat),n-1))
    reject='拒绝' if p_val<alpha else '接受'
    in_ci='在区间内' if ci[0]<=mu0<=ci[1] else '不在区间内'
    print('H0: mu='+str(mu0)+'  '+reject+'  ('+in_ci+')')
```
