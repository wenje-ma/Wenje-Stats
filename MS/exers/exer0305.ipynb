## 030504

设 $X_1,X_2,\cdots,X_n$ 是来自两点分布总体 $B(1,p)$ 的样本, 其中 $0<p<1$ 为未知参数.

(1)

求 $p^m$ 的一致最小方差无偏估计, 其中 $m>0$ 为自然数.

(2)

求 $p(1-p)$ 的一致最小方差无偏估计.

### 030504-解答

(1)

两点分布的联合分布列为

$$
\begin{aligned}
&P\{X_1=x_1,X_2=x_2,\cdots,X_n=x_n\}\\
&=p^{\sum_{i=1}^nx_i}(1-p)^{n-\sum_{i=1}^nx_i}\\
&=\exp\{n\ln(1-p)\}\exp\left\{\ln\frac p{1-p}\sum_{i=1}^nx_i\right\},
\end{aligned}
$$

令 $\varphi=\ln p/(1-p)$, 则 $1-p=1/(e^\varphi+1)$,

$$
\begin{aligned}
&P\{X_1=x_1,X_2=x_2,\cdots,X_n=x_n\}\\
&=\exp\left\{\frac n{e^\varphi+1}\right\}\exp\left\{\varphi\sum_{i=1}^nx_i\right\},
\end{aligned}
$$

故其充分完备统计量为 $T=\sum_{i=1}^nX_i$. 取简单的无偏估计: $\widehat{g}=\prod_{i=1}^m X_i$, 显然

$$
E(\widehat{g})=\prod_{i=1}^m EX_i=p^m.
$$

由于 $\widehat{g}$ 并不是 $T$ 的函数, 因此计算条件期望 $E(X_1,X_2,\cdots,X_m|T=t)$:

- 当 $t < m$ 时, $X_i$ 至少有一个为 $0$, 因此期望为 $0$;
- 当 $t \ge m$ 时,

$$
\begin{aligned}
&E(X_1,X_2,\cdots,X_n|T=t)\\
&=\frac{P(X_1=1,X_2=1,\cdots,X_m=1, T=t)}{P(T=t)}\\
&=\frac{p^m\binom{n-m}{t-m} p^{t-m} (1-p)^{n-t}}{\binom{n}{t} p^t (1-p)^{n-t}}\\
&=\left.\binom{n-m}{t-m}\middle/\binom{n}{t}\right..
\end{aligned}
$$

因此 $p^n$ 的一致最小方差无偏估计 为

$$
\left.\binom{n-m}{T-m}\middle/\binom{n}{T}\right.I_{\{T=\sum_{i=1}^nX_i=m,m+1,s\}}.
$$

(2)

由于两点分布的方差为 $p(1-p)$, 样本方差

$$
S^2=\frac{1}{n-1}\sum_{i=1}^n (X_i-\overline{X})^2
$$

本身就是总体方差的无偏估计, 因此将 $S^2$ 用 $T=\sum_{i=1}^nX_i$ 表示即可: 依据 Bernoulli 变量独有的 $X_i^2=X_i=\{0,1\}$,

$$
\begin{aligned}
S^2&=\frac{1}{n-1}\sum_{i=1}^n (X_i-\overline{X})^2\\
&=\frac{1}{n-1}\left(\sum_{i=1}^n X_i^2-n\overline{X}^2\right)\\
&=\frac{1}{n-1}\left(\sum_{i=1}^n X_i-\frac{1}{n}(\sum X_i)^2\right)\\
&=\frac{1}{n-1}\left(T-\frac{T^2}{n}\right)=\frac{T(n-T)}{n(n-1)},
\end{aligned}
$$

其中 $T=\sum_{i=1}^nX_i=1,2,\cdots,n$.

## 030506

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(0,\sigma^2)$ 的样本, 其中 $\sigma>0$ 为未知参数, 分别求 $\sigma$ 和 $\sigma^4$ 的一致最小方差无偏估计.

### 030506-解答

正态分布的联合分布列为

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\sigma^2)\\
&=(2\pi\sigma^2)^{-n/2}\exp\left\{-\frac1{2\sigma^2}\sum_{i=1}^nX_i^2\right\},
\end{aligned}
$$

令 $\varphi=-1/(2\sigma^2)$, 则

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\sigma^2)\\
&=(-\varphi/\pi)^{n/2}\exp\left\{\varphi\sum_{i=1}^nX_i^2\right\},
\end{aligned}
$$

则正态分布的充分完备统计量为 $T=\sum_{i=1}^nX_i^2$, 已知

$$
\frac{\sum_{i=1}^nX_i^2}{\sigma^2}\sim\chi^2(n),
$$

- 求 $\sigma$ 的一致最小方差无偏估计 时, 自然地考虑 $\sqrt{T}$:

令 $U=T/\sigma^2\sim\chi^2(n)$, 则

$$
E(\sqrt U)=\int_0^{+\infty}\sqrt u\frac{u^{n/2-1}e^{-u/2}}{2^{n/2}\Gamma(n/2)}du=\frac{\sqrt2\Gamma((n+1)/2)}{\Gamma(n/2)}.
$$

因此

$$
\sigma=\frac{E(\sqrt T)}{E(\sqrt U)}=E\left(\frac{\Gamma(n/2)\sqrt T}{\sqrt2\Gamma((n+1)/2)}\right),
$$

即

$$
\frac{\Gamma(n/2)\sqrt T}{\sqrt2\Gamma((n+1)/2)}
$$

是 $\sigma$ 的一致最小方差无偏估计.

- 求 $\sigma^4$ 的一致最小方差无偏估计 时, 自然地考虑 $T^2$:

令 $U=T/\sigma^2\sim\chi^2(n)$, 则

$$
E(U^2)=\frac{E(T^2)}{\sigma^4}
$$

已知 $E(U)=n$, $D(U)=2n$, 则

$$
\sigma^4=E\left(\frac{T^2}{E(U^2)}\right)=E\left(\frac{T^2}{n(n+2)}\right),
$$

因此

$$
\frac{T^2}{n(n+2)}
$$

是 $\sigma^4$ 的一致最小方差无偏估计.

## 030507

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本, 其中 $\mu\in\mathbb{R},\sigma>0$ 都是未知参数.

(1)

分别求 $\mu+\sigma^2$ 和 $\mu^2/\sigma^2$ 的一致最小方差无偏估计;

(2)

证明 $\widehat{\mu}^2=\overline{X}^2-S^2/n$ 是 $\mu^2$ 的一致最小方差无偏估计.

### 030507-解答

(1)

已知 $\overline{X}$ 是 $\mu$ 的一致最小方差无偏估计, $S^2$ 是 $\sigma^2$ 的一致最小方差无偏估计, 因此 $\mu+\sigma^2$ 的一致最小方差无偏估计 为

$$
\overline{X} + S^2.
$$

由于 $\overline{X}$ 与 $S^2$ 独立, 因此考虑

$$
E\left(\frac{\overline{X}^2}{S^2}\right)=E(\overline{X}^2)E(1/S^2).
$$

已知 $\overline{X}\sim N(\mu,\sigma^2/n)$, 则

$$
E(\overline{X}^2)=D(\overline{X}) + (E\overline{X})^2=\frac{\sigma^2}{n} + \mu^2.
$$

已知 $U=(n-1)S^2/\sigma^{2}\sim\chi^2(n-1)$, 因此

$$
E(1/S^2)=E\left(\frac{n-1}{\sigma^2U}\right)=\frac{n-1}{\sigma^2}E(1/U),
$$

$$
\begin{aligned}
E(1/U)&=\int_0^\infty1/uf(u)du\\
&=\frac{\Gamma(k/2-1)}{2\Gamma(k/2)}=\frac1{k-2},
\end{aligned}
$$

因此 $k=n-1$ 时 $E(1/S^2)=(n-1)/(\sigma^2(n-3))$. 即

$$
\begin{aligned}
E\left(\frac{\overline{X}^2}{S^2}\right)&=\left(\frac{\sigma^2}{n}+\mu^2\right)\frac{n-1}{\sigma^2(n-3)}\\
&=\frac{n-1}{n(n-3)}+\frac{\mu^2(n-1)}{\sigma^2(n-3)}.
\end{aligned}
$$

因此

$$
\frac{(n-3)(\sum X_i)^2}{n\sum(X_i-\overline X)^2}-\frac1n
$$

为 $\mu^{2}/\sigma^{2}$ 的一致最小方差无偏估计.

(2)

无偏性:

$$
\begin{aligned}
&E\left(\overline{X}^2-\frac{1}{n}S^2\right)\\
&=E\overline{X}^2-\frac{1}{n}E(S^2)\\
&=\mu^2 + \frac{\sigma^2}{n}-\frac{\sigma^2}{n}=\mu^2
\end{aligned}
$$

故的确是无偏估计.

充分完备性: $\overline{X}^2-\frac{1}{n}S^2$ 是 $(\overline{X}, S^2)$ 的函数易证.
