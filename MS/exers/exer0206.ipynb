## 020601

设 $X_1,X_2,\cdots,X_n$ 是来自 Poisson 分布总体 $P(\lambda)$, $\lambda>0$ 的样本, 证明 $\sum_{i=1}^nX_i$ 为完备统计量.

### 020601-解答

Poisson 分布可表示为自然形式的指数族:

$$
\begin{aligned}
&P(X_1=x_1,X_2=x_2,\cdots,X_n=x_n;\lambda)\\
&=e^{-n\lambda}\exp\left\{\sum_{i=1}^n\ln\lambda x_i\right\}\prod_{i=1}^n\frac{1}{x_i!}
\end{aligned}
$$

令 $\varphi=\ln\lambda$, 则 $\lambda=e^\varphi$,

$$
\begin{aligned}
&P(X_1=x_1,X_2=x_2,\cdots,X_n=x_n;\varphi)\\
&=e^{-ne^\varphi}\exp\left\{\sum_{i=1}^n\varphi x_i\right\}\prod_{i=1}^n\frac{1}{x_i!}
\end{aligned}
$$

由于 $\lambda>0$, 则 $\varphi\in\mathbb R$, 自然参数空间 $\Theta^\varphi=\mathbb R$ 含内点, 故 $T=\sum_{i=1}^nX_i$ 为完备统计量.

## 020605

设 $X_1,X_2,\cdots,X_n$ 是来自 Rayleigh 分布的样本,其密度函数为

$$
f(x)=
\begin{cases}
2\theta xe^{-\theta x^2},&x>0,\\
0,&\text{Others},
\end{cases}
$$

其中 $\theta>0$. 试给出一个完备统计量.

### 020605-解答

Rayleigh 分布可表示为自然形式的指数族:

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\theta)\\
&=(2\theta)^n\exp\left\{-\theta\sum_{i=1}^nx_i^2\right\}\prod_{i=1}^nx_iI_{\{x_i>0\}}
\end{aligned}
$$

令 $\varphi=-\theta$, 则 $\theta=-\varphi$,

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\theta)\\
&=(-2\varphi)^n\exp\left\{\varphi\sum_{i=1}^nx_i^2\right\}\prod_{i=1}^nx_iI_{\{x_i>0\}}
\end{aligned}
$$

由于 $\theta>0$, 则 $\varphi<0$, 参数空间 $\Theta^*=(-\infty,0)$ 含内点, 则其充分统计量 $T=\sum_{i=1}^nX_i^2$ 完备.

## 020606

设 $X_1,X_2,\cdots,X_n$ 是来自指数分布 $E(\lambda)$ 的样本, 其密度函数为

$$
f(x)=\begin{cases}
\lambda e^{-\lambda x},&x>0,\\
0,&x\le0,
\end{cases}
$$

其中 $\lambda>0$. 证明 $\sum_{i=1}^nX_i$ 为完备统计量.

### 020606-解答

指数分布可表示为自然形式的指数族:

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n,\lambda)\\
&=\lambda^n\exp\left\{\sum_{i=1}^n(-\lambda)x_i\right\}
\end{aligned}
$$

令 $\varphi=-\lambda$, 则

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n,\lambda)\\
&=(-\varphi)^n\exp\left\{\sum_{i=1}^n\varphi x_i\right\}
\end{aligned}
$$

由于 $\lambda>0$, 则 $\varphi<0$, 参数空间 $\Theta^*=(-\infty,0)$ 含内点, 故 $\sum_{i=1}^nX_i$ 完备.

## 020607

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本.

(1)

在 $\mu$ 已知时, 给出 $\sigma^2$ 的一个完备统计量；

(2)

在 $\sigma^2$ 已知时, 给出 $\mu$ 的一个完备统计量.

### 020607-解答

(1)

正态分布可表示为自然形式的指数族:

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\mu,\sigma)\\
&=(2\pi\sigma^2)^{-n/2}\exp\left\{\sum_{i=1}^n(-\frac1{2\sigma^2})(x_i-\mu)^2\right\}.
\end{aligned}
$$

令 $\varphi=-1/2\sigma^2$, 则 $\sigma^2=-1/2\varphi$,

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\varphi)\\
&=(-\varphi/\pi)^{n/2}\exp\left\{\sum_{i=1}^n\varphi(x_i-\mu)^2\right\}.
\end{aligned}
$$

参数空间 $\Theta^*=(-\infty,0)$ 含内点, 则其充分统计量 $T=\sum_{i=1}^n(X_i-\mu)^2$ 完备.

(2)

对于

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\mu,\sigma)\\
&=(2\pi\sigma^2)^{-n/2}\exp\left\{\sum_{i=1}^n\left(-\frac1{2\sigma^2}\right)(x_i-\mu)^2\right\}\\
&=(2\pi\sigma^2)^{-n/2}\exp\left\{-\sum_{i=1}^n\frac{x_i^2}{2\sigma^2}+\sum_{i=1}^n\frac{\mu x_i}{\sigma^2}-\sum_{i=1}^n\frac{\mu^2}{2\sigma^2}\right\}\\
&=(2\pi\sigma^2)^{-n/2}\exp\left\{\sum_{i=1}^n\frac{\mu x_i}{\sigma^2}-\sum_{i=1}^n\frac{\mu^2}{2\sigma^2}\right\}\exp\left\{-\sum_{i=1}^n\frac{x_i^2}{2\sigma^2}\right\}.
\end{aligned}
$$

令 $\varphi=\mu/\sigma^2$, 则 $\mu^2=\varphi^2\sigma^4$,

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\varphi)\\
&=(2\pi\sigma^2)^{-n/2}\exp\left\{\sum_{i=1}^n\varphi x_i-\frac{n\varphi^2\sigma^2}2\right\}\exp\left\{-\sum_{i=1}^n\frac{x_i^2}{2\sigma^2}\right\}.
\end{aligned}
$$

$\mu\in\mathbb R$, $\varphi=\mu/\sigma^2\in\mathbb R$, 参数空间 $\Theta^*=\mathbb R$ 含内点, 则其充分统计量 $T=\sum_{i=1}^nX_i$ 完备.

## 020610

设 $X_1,X_2,\cdots,X_n$ 是来自均匀分布总体 $U[\theta,2\theta]$, $\theta>0$ 的样本, 证明 $(X_{(1)},X_{(n)})$ 是充分统计量但不是完备统计量.

### 020610-解答

$X_{(1)}$ 与 $X_{(n)}$ 联合密度:

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\theta)\\
&=\frac{1}{\theta^n}I_{\{x_{(1)}\ge\theta\}}I_{\{x_{(n)}\le2\theta\}}
\end{aligned}
$$

由因子分解定理, $(X_{(1)},X_{(n)})$ 为充分统计量.

由于均匀分布的联合密度函数并不是指数族, 因此只能用定义判断完备统计量.

由尺度不变性, 令 $Z_i=X_i/\theta\sim U[1,2]$, 则

$$
\frac{X_{(n)}}{X_{(1)}}=\frac{\theta Z_{(n)}}{\theta Z_{(1)}}=\frac{Z_{(n)}}{Z_{(1)}},
$$

其分布与 $\theta$ 无关. 令

$$
\varphi(T)=\frac{X_{(n)}}{X_{(1)}}-E\left(\frac{X_{(n)}}{X_{(1)}}\right),
$$

则 $\varphi(T)$ 仅为 $T=(X_{(1)},X_{(n)})$ 的函数, $E_\theta(\varphi(T))=0$ 但 $\varphi(T)\not\equiv0$.
