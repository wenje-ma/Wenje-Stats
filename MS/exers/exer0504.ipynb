# 0504

## 050401

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本, 其中 $\mu\in\mathbb R$, $\sigma^2>0$ 均未知, 利用假设检验方法导出 $\sigma^2$ 的置信水平为 $1-\alpha$ 的置信区间和置信上下限.

### 050401-解答

构造检验统计量:

$$
Q=\frac{(n-1)S^2}{\sigma^2} \sim \chi^2(n-1),
$$

依据对偶性, 其置信水平为 $1-\alpha$ 的置信区间满足

$$
P\left\{\chi^2_{\alpha/2}(n-1) \le \frac{(n-1)S^2}{\sigma^2} \le \chi^2_{1-\alpha/2}(n-1)\right\}=1-\alpha.
$$

解得置信区间为

$$
\sigma^2 \in \left[\frac{(n-1)S^2}{\chi^2_{1-\alpha/2}(n-1)},\ \frac{(n-1)S^2}{\chi^2_{\alpha/2}(n-1)}\right].
$$

置信上限为

$$
\frac{(n-1)S^2}{\chi^2_{\alpha}(n-1)},
$$

置信下限为

$$
\frac{(n-1)S^2}{\chi^2_{1-\alpha}(n-1)}.
$$

## 050402

设总体 $X_i\sim N(\mu_1,\sigma_1^2)$, $i=1,2,\cdots,m$ 是来自总体 $X$ 的样本, 总体 $Y_i\sim N(\mu_2,\sigma_2^2)$, $i=1,2,\cdots,m$ 是来自总体 $Y$ 的样本, 且两个样本相互独立, 其中

$$
\begin{aligned}
\mu_1\in\mathbb R,\quad&\sigma_1>0,\\
\mu_2\in\mathbb R,\quad&\sigma_2>0\\
\end{aligned}
$$

均未知, 利用假设检验方法求出 $\sigma_1^2/\sigma_2^2$ 的置信水平为 $1-\alpha$ 的置信区间和置信上下限.

### 050402-解答

构造检验统计量:

$$
Q=\frac{S_1^2/\sigma_1^2}{S_2^2/\sigma_2^2} \sim F(m-1,n-1),
$$

依据对偶性, 其置信水平为 $1-\alpha$ 的置信区间满足

$$
P\{F_{\alpha/2}(m-1,n-1) \le \frac{S_1^2/\sigma_1^2}{S_2^2/\sigma_2^2} \le F_{1-\alpha/2}(m-1,n-1)\}=1-\alpha.
$$

解得置信区间为

$$
\frac{\sigma_1^2}{\sigma_2^2} \in \left[\frac{S_1^2}{S_2^2}F_{\alpha/2}(n-1,m-1),\ \frac{S_1^2}{S_2^2}F_{1-\alpha/2}(n-1,m-1)\right].
$$

因此置信上限为

$$
\frac{S_1^2}{S_2^2}F_{1-\alpha}(n-1,m-1),
$$

置信下限为

$$
\frac{S_1^2}{S_2^2}F_{\alpha}(n-1,m-1).
$$
