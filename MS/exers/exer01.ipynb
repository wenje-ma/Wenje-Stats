## 0104

设 $X_1,X_2,\cdots,X_n$ 是来自总体 $X$ 的样本, 在下列三种情况下, 分别写出样本 $X_1,X_2,\cdots,X_n$ 的联合分布列或联合密度函数.

(1)

总体 $X$ 服从几何分布, 其分布列为

$$
P\{X=x\}=(1-p)^{x-1}p,
$$

其中 $0<p<1$, $x=1,2,\cdots$.

(2)

总体 $X$ 服从指数分布, 其密度函数为

$$
f(x)=
\begin{cases}
\lambda e^{-\lambda x},x>0,\\
0,\text{Others},
\end{cases}
$$

其中 $\lambda>0$.

(3)

总体 $X$ 的密度函数为

$$
f(x)=\frac{\lambda}{2}e^{-\lambda|x|},
$$

其中 $x\in\mathbb{R}$, $\lambda>0$.

### 0104-解答

(1)

$$
p(x_{1},x_{2},\cdots,x_{n})=(1-p)^{\sum_{i=1}^nx_i-n}p^n.
$$

(2)

$$
f(x_{1},x_{2},\cdots,x_{n})=\lambda^ne^{-\lambda\sum_{i=1}^nx_i}I_{\{x_i>0\}},
$$

其中 $i=1,2,\cdots,n$.

(3)

$$
f(x_{1},x_{2},\cdots,x_{n})=\frac{\lambda^n}{2^n}e^{-\lambda\sum_{i=1}^n|x_i|},
$$

其中 $i=1,2,\cdots,n$.

##  0106

设 $X_1, X_2, \cdots, X_n$ 是来自总体 $X$ 的样本, 写出两点分布, 二项分布, Poisson 分布, 几何分布, 均匀分布, 指数分布, 正态分布的样本分布族.

### 0106-解答

- 两点分布:

$$
\mathcal{F}=\{p^{\sum x_i}(1-p)^{n-\sum x_i}:0<p<1\},
$$

- 二项分布:

$$
\mathcal{F}=\{\prod_{i=1}^n\binom{m}{x_i}p^{\sum x_i}(1-p)^{mn-\sum x_i}:0<p<1\};
$$

- Poisson 分布:

$$
\mathcal{F}=\{\frac{\lambda^{\sum_{i=1}^nx_i}}{\prod_{i=1}^nx_i!}e^{-n\lambda}:\lambda>0\},
$$

- 几何分布:

$$
\mathcal{F}=\{p^n(1-p)^{\sum_{i=1}^nx_i-n}:0<p<1\},
$$

- 均匀分布:

$$
\mathcal{F}=\{\frac{1}{(b-a)^n}I_{\{a< x_i< b\}}:-\infty<a<b<+\infty\},
$$

- 指数分布:

$$
\mathcal{F}=\{\lambda^ne^{-\lambda\sum_{i=1}^nx_i}I_{x_i>0}:\lambda>0\},
$$

- 正态分布:

$$
\mathcal{F}=\{\frac{1}{(\sqrt{2\pi}\sigma)^n}\exp\{-\frac{1}{2\sigma^2}\sum_{i=1}^n(x_i-\mu)^2\}:\mu\in\mathbb R,\sigma>0\}.
$$

## 0110

设 $F(x)$ 为总体 $X$ 的分布函数, $X_1,X_2,\cdots,X_n$ 是来自总体 $X$ 的样本, $F_n(x)$ 为其经验分布函数. 证明对任意给定实数   $x$, 有:

(1)

$$
E(F_n(x))=F(x);
$$

(2)

$$
D(F_n(x))=\frac{1}{n}F(x)(1-F(x)).
$$

### 0110-解答

(1)

由经验分布函数的定义, 期望的线性性, 同分布的性质 $X_1=X_2=s=X_n=X$, 示性函数 $I_A$ 的数学期望,

$$
\begin{aligned}
E(F_n(x))&=E(\frac1n\sum_{i=1}^nI_{\{X_i\le x\}})\\
&=\frac{1}{n}\sum_{i=1}^{n}E(I_{\{X_i\le x\}})\\
&=\frac{1}{n}nE(I_{\{X\le x\}})=F(x).
\end{aligned}
$$

(2)

由经验分布函数的定义, 方差数乘公式, 独立随机变量方差和等于各方差之和, 同分布的性质 $X_1=X_2=s=X_n=X$, 方差的定义, 示性函数 $I_A$ 的数学期望, Bernoulli 分布的期望 $EX=p$, 方差 $DX=p(1-p)$,

$$
\begin{aligned}
D(F_n(x))&=D(\frac1n\sum_{i=1}^{n}I_{\{X_i\le x\}})\\
&=\frac{1}{n^2}D(\sum_{i=1}^{n}I_{\{X_i\le x\}})\\
&=\frac{1}{n^2}nD(I_{\{X\le x\}})\\
&=\frac{1}{n}E(I_{\{X_i\le x\}})(1-E(I_{\{X_i\le x\}}))\\
&=\frac{1}{n}F(x)(1-F(x)).
\end{aligned}
$$
