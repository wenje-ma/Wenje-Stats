## 050105

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,1)$ 的样本, 其中 $\mu\in\mathbb{R}$ 为未知参数, 设检验问题

$$
H_0: \mu=0;\quad H_1: \mu=1
$$

的拒绝域为 $D=\{\overline{X}\ge0.98\}$.

(1)

若 $n=4$, 求该检验犯第一类错误的概率和犯第二类错误的概率;

(2)

若要使该检验犯第一类错误的概率不超过 $0.01$, 样本容量 $n$ 最少取多少?

### 050105-解答

(1)

$$
\overline{X} \sim N\left(\mu, \frac{1}{n}\right),
$$

标准化:

$$
\begin{aligned}
\varphi(\mu)&=1-P\{\overline{X} < 0.98\}\\
&=1-P\{\sqrt n(\overline{X}-\mu)<\sqrt{n}(0.98-\mu)\}\\
&=1-\Phi(\sqrt{n}(0.98-\mu)).
\end{aligned}
$$

当 $n=4$ 时, 第一类错误概率:

$$
\begin{aligned}
\alpha&=P\{\overline X\ge0.98|H_0:\mu=0\}\\
&=1-\Phi(1.96)=0.025.
\end{aligned}
$$

第二类错误概率:

$$
\begin{aligned}
\beta&=P\{\overline{X} < 0.98|H_1: \mu=1\}\\
&=\Phi(-0.04)=0.484.
\end{aligned}
$$

(2)

要使第一类错误概率不超过 $0.01$:

$$
\begin{aligned}
1-\Phi(\sqrt{n}\times 0.98)&\le0.01\\
\Phi(\sqrt{n}\times 0.98)&\ge0.99,
\end{aligned}
$$

查表得 $\Phi(2.33)=0.99$, 故

$$
\begin{aligned}
\sqrt{n}\times0.98&\ge2.33\\
n&\ge\left\lceil(\frac{2.33}{0.98})^2\right\rceil=6.
\end{aligned}
$$

## 050107

设样本 $X$ (容量为 $1$) 取自具有密度函数 $f(x)$ 的总体, 关于总体的两个假设如下:

$$
H_0:f(x)=
\begin{cases}1,&0<x<1,\\
0,&\text{Others}.
\end{cases}
$$

$$
H_1:f(x)=
\begin{cases}
2x,&0<x<1,\\
0,&\text{Others}.
\end{cases}
$$

设该检验的拒绝域为 $D=\{X>2/3\}$. 求该检验犯第一类错误的概率和犯第二类错误的概率.

### 050107-解答

$$
\varphi(x)=P\left\{X>\frac23\right\},
$$

$$
\begin{aligned}
\alpha&=\left\{X>\frac23|H_0\right\}\\
&=\int_{2/3}^11dx=\frac13,
\end{aligned}
$$

$$
\begin{aligned}
\beta&=\left\{X\le\frac23|H_1\right\}\\
&=\int_0^{2/3}2xdx=\frac49.
\end{aligned}
$$

## 050108

设 $X_1,X_2,\cdots,X_{10}$ 是来自均匀分布总体 $U[0,\theta]$ 的样本, 其中 $\theta>0$ 为未知参数, 设检验问题

$$
H_0: \theta=1;\quad H_1: \theta>1
$$

的拒绝域为

$$
D=\{X_{(10)}\ge0.95\}.
$$

(1)

求该检验的显著性水平;

(2)

求该检验的功效函数.

### 050108-解答

(1)

显著性水平是犯第一类错误的最大概率, 此时 $H_0: \theta=1$ 为真, $X_{(10)}$ 的分布函数:

$$
\begin{aligned}
F_{X_{(10)}}(x)&=P\{X_{(10)}\le x\}\\
&=P\{X_1\le x,X_2\le x,\cdots,X_{10}\le x\}\\
&=
\begin{cases}
0,&x<0,\\
(x/\theta)^{10},&0\le x\le \theta,\\
1,&x>\theta.
\end{cases}
\end{aligned}
$$

当 $\theta=1$ 时,

$$
\begin{aligned}
\alpha&=P\{X_{(10)} \ge 0.95 \mid \theta=1\}\\
&=1-F_{X_{(10)}}(0.95)=1-0.95^{10}.
\end{aligned}
$$

(2)

功效函数

$$
\varphi(\theta)=P\{X_{(10)} \ge 0.95\},
$$

当 $\theta < 0.95$ 时, $X_{(10)} \le \theta < 0.95$, 故

$$
\varphi(\theta)=P\{X_{(10)} \ge 0.95\}=0.
$$

当 $\theta \ge 0.95$ 时:

$$
\begin{aligned}
\varphi(\theta)&=1-P\{X_{(10)}\le0.95\}\\
&=1-F_{X_{(10)}}(0.95)\\
&=1-\left(\frac{0.95}{\theta}\right)^{10},
\end{aligned}
$$

因此功效函数:

$$
\varphi(\theta)=
\begin{cases}
0,&\theta<0.95,\\
1-(0.95/\theta)^{10},&\theta\ge0.95.\end{cases}
$$
