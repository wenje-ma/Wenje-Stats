## 020201

设 $X_1,X_2,X_3,X_4,X_5$ 是来自总体 $X$ 的样本, 且 $X\sim N(0,1)$.

(1) 确定 $a,b$, 使得

$$
a(X_1-2X_2+3X_3)^2+b(4X_4-5X_5)^2
$$

服从 $\chi^2$ 分布, 并指出自由度;

(2) 确定 $c$ 使得

$$
\frac{c\sum_{i=1}^3X_i}{\sqrt{\sum_{i=4}^5X_i^2}}
$$

服从 $t$ 分布, 并指出自由度;

(3) 确定 $d$ 使得

$$
\frac{d\sum_{i=1}^3X_i^2}{\sum_{i=4}^5X_i^2}
$$

服从 $F$ 分布, 并指出自由度.

### 020201-解答

(1)

由于 $X_1, X_2, \cdots, X_n \sim N(0, 1)$ 独立, 则正态分布具有可加性, 因此

$$
X_1-2X_2+3X_3\sim N(0,14)
$$

$$
4X_4-5X_5\sim N(0,41).
$$

由于

$$
\frac1{14}(X_1-2X_2+3X_3)^2+\frac1{41}(4X_4-5X_5)^2\sim\chi^2(2),
$$

因此 $a=1/14$, $b=1/41$, 自由度为 $2$.

(2)

由于 $X_1, X_2, \cdots, X_n \sim N(0, 1)$ 独立, 则正态分布具有可加性, 因此

$$
\sum_{i=1}^3X_i\sim N(0,3).
$$

依据 $\chi^2$ 分布的定义,

$$
\sum_{i=4}^5X_i^2\sim\chi^2(2).
$$

由于

$$
\frac{\pm\sqrt{2/3}\sum_{i=1}^5X_i}{\sqrt{\sum_{i=4}^5X_i^2}}\sim t(2),
$$

因此 $c=\pm\sqrt{2/3}$, 自由度为 $2$.

(3) 依据 $\chi^2$ 分布的定义,

$$
\sum_{i=1}^{3}X_i^2\sim \chi^2(3),
$$

$$
\sum_{i=4}^{5}X_i^2\sim \chi^2(2).
$$

由于

$$
\frac{\sum_{i=1}^{3}X_i^2/3}{\sum_{i=4}^{5}X_i^2/2}\sim F(3,2),
$$

则 $d=2/3$, 第一自由度为 $3$, 第二自由度为 $2$.

## 020202

设随机变量 $X$ 服从自由度为 $n$ 的 $\chi^2$ 分布, 证明:

$$
E(X^k)=\frac{2^k\Gamma(n/2+k)}{\Gamma(n/2)},
$$

其中 $k=1,2,\cdots$. 特别地, 有 $E(X)=n$, $D(X)=2n$​, 已知 $\chi^2$ 分布的密度函数

$$
f(x)=\frac{x^{n/2-1}e^{-x/2}}{2^{n/2}\Gamma(n/2)}.
$$

### 020202-解答

$$
EX^k=\int_{0}^{+\infty}x^kf(x)dx
=\int_{0}^{+\infty}x^k\frac{x^{n/2-1}e^{-x/2}}{2^{n/2}\Gamma(n/2)}dx.
$$

为了凑出 Gamma 函数, 令 $t=x/2$, 则 $dx=2dt$,

$$
EX^k=\frac{2^k}{\Gamma(n/2)}\int_{0}^{+\infty}t^{n/2+k-1}e^{-t}dt=\frac{2^k\Gamma(n/2+k)}{\Gamma(n/2)},
$$

$$
EX=\frac{2\Gamma(n/2+1)}{\Gamma(n/2)}=n,
$$

$$
EX^2=\frac{4\Gamma(n/2+2)}{\Gamma(n/2)}=n^2+2n,
$$

$$
DX=EX^2-(EX)^2=2n.
$$

## 020203

设某连续型总体 $X$ 的分布函数 $F(x)$ 是连续的严格单调递增函数, $X_1,X_2,\cdots,X_n$ 是来自总体 $X$ 的样本, 证明:

(1)

$$
Y=F(X)\sim U(0,1).
$$

(2)

$$
U=-2\sum_{i=1}^n\ln Y_i\sim\chi^2(2n)​.
$$

### 020203-解答

(1)

由于 $F(x)$ 严格单调, 则 $F^{-1}(y)$ 存在.

依据分布函数的定义,

$$
P(Y\le y)=P(X\le F^{-1}(y))=F(F^{-1}(y))=y\in(0,1),
$$

故 $Y\sim U(0,1)$.

(2)

由于 $\chi^2$ 分布的再生性, 同分布的性质 $Y_1=Y_2=s=Y_n=Y$, 只需证明

$$
-2\ln Y\sim\chi^2(2).
$$

由于

$$
P(-2\ln Y\le y)=P(Y\ge e^{-y/2})=1-e^{-y/2},
$$

则

$$
-2\ln Y\sim E(1/2).
$$

由于数学期望为 $2$ 的指数分布是 $\chi^2(2)$ 分布, 因此

$$
-2\ln Y_1\sim \chi^2(2).
$$

## 020206

设随机变量 $X$ 服从参数为 $\alpha,\lambda$ 的 Gamma 分布 $Ga(\alpha,\lambda),\alpha>0,\lambda>0$, 密度函数为

$$
f(x)=\begin{cases}
\lambda^\alpha x^{\alpha-1}e^{-\lambda x}/\Gamma(\alpha),&x>0,\\
0,&x\le0,
\end{cases}
$$

证明:

(1)

$X$ 的特征函数为

$$
\varphi(t)=\left(1-\frac{it}{\lambda}\right)^{-\alpha}.
$$

(2)

$$
E(X^k)=\frac{\Gamma(\alpha+k)}{\Gamma(\alpha)}\cdot\frac{1}{\lambda^k},
$$

特别地, $E(X)=\alpha/\lambda,D(X)=\alpha/\lambda^2$;

(3)

如果 $X_i\sim\text{Ga}(\alpha_i,\lambda),i=1,2,\cdots,n$ 且 $X_1,X_2,\cdots,X_n$ 相互独立, 则

$$
\sum_{i=1}^nX_i\sim\text{Ga}(\alpha_1+\alpha_2+\cdots+\alpha_n,\lambda);
$$

(4)

若取 $\alpha=n/2,\lambda=1/2$, 则 $\text{Ga}(n/2,1/2)$ 是 $\chi^2(n)$ 分布;

(5)

若取 $\alpha=1$, 则 $\text{Ga}(1,\lambda)$ ​是指数分布.

### 020206-解答

(1)

由特征函数的定义,

$$
\varphi(t)=Ee^{itX}=\int_{0}^{+\infty}e^{itx}\frac{1}{\Gamma(\alpha)}\lambda^\alpha x^{\alpha-1}e^{-\lambda x}dx,
$$

为了凑 Gamma 函数, 令 $u=(\lambda-it)x$, 则

$$
x=\frac{u}{\lambda-it},\quad dx=(\lambda-it)^{-1}du,
$$

因此

$$
\varphi(t)=\left(1-\frac{it}{\lambda}\right)^{-\alpha}.
$$

(2)

$$
EX^k=\int_{0}^{+\infty}x^k\frac{1}{\Gamma(\alpha)}\lambda^\alpha x^{\alpha-1}e^{-\lambda x}dx,
$$

为了凑 Gamma 函数, 令 $t=\lambda x$, 则 $dt=\lambda dx$,

$$
EX^k=\frac{1}{\Gamma(\alpha)\lambda^k}\int_{0}^{+\infty}t^{\alpha+k-1}e^{-t}dt=\frac{\Gamma(\alpha+k)}{\Gamma(\alpha)\lambda^k},
$$

$$
EX=\frac{\Gamma(\alpha+1)}{\lambda\Gamma(\alpha)}=\frac{\alpha}{\lambda},
$$

$$
EX^2=\frac{\Gamma(\alpha+2)}{\lambda^2\Gamma(\alpha)}=\frac{\alpha(\alpha+1)}{\lambda^2},
$$

$$
DX=EX^2-(EX)^2=\frac{\alpha}{\lambda^2}.
$$

(3)

由于 $X_1,X_2,\cdots,X_n$ 独立, 则利用特征函数:

$$
\varphi_Y(t)=\prod_{i=1}^n\varphi_{X_i}(t)=\left(1-\frac{it}{\lambda}\right)^{-\sum_{i=1}^n\alpha_i},
$$

则 $\sum_{i=1}^nX_i\sim\text{Ga}(\sum_{i=1}^n\alpha_i,\lambda)$.

(4)

取 $\alpha=n/2$, $\lambda=1/2$, 则密度函数变为

$$
f_1(x)=
\begin{cases}
2^{n/2}x^{n/2-1}e^{-x/2}/\Gamma(n/2),&x\ge0,\\
0,&x<0,
\end{cases}
$$

则 $X\sim\chi^2(n)$.

(5) 取 $\alpha=1$, 则密度函数变为

$$
f_2(x)=
\begin{cases}
\lambda e^{-\lambda x},&x>0,\\
0,&x\le0,
\end{cases}
$$

则 $X\sim E(\lambda)$.

## 020207

设随机变量 $X\sim\chi^2(m)$, $Y\sim\chi^2(n)$, 且 $X$ 与 $Y$ 相互独立, 证明 $U=X/Y$与$V=X+Y$ ​相互独立, 已知 $\chi^2$ 分布的密度函数

$$
f(x)=\frac{x^{n/2-1}e^{-x/2}}{2^{n/2}\Gamma(n/2)}.
$$

### 020207-解答

由于 $X,Y$ 独立, 则其联合密度为

$$
f_{X,Y}(x,y)=f_X(x)f_Y(y)=\frac{x^{m/2-1}y^{n/2-1}}{2^{(m+n)/2}\Gamma(m/2)\Gamma(n/2)}e^{-(x+y)/2}.
$$

令

$$
\begin{cases}
U=X/Y,\\V=X+Y,
\end{cases}
$$

则

$$
\begin{cases}
X=UV/(1+U),\\
Y=V/(1+U),
\end{cases}
$$

$$
J=\frac{\partial(x,y)}{\partial(u,v)}=\frac{V}{(1+U)^2}.
$$

欲证 $U$ 与 $V$ 独立, 只需证其联合密度函数能拆成两个单变量函数乘积, 即

$$
f_{U,V}(u,v)=g(u)h(v),
$$

因此

$$
\begin{aligned}
f_{U,V}(u,v)&=f_{X,Y}\left(\frac{uv}{1+u},\frac{v}{1+u}\right)|J|\\
&=\left(\frac{\Gamma((m+n)/2)}{\Gamma(m/2)\Gamma(n/2)}\frac{u^{m/2-1}}{(1+u)^{(m+n)/2}}\right)\\
&\cdot\left(\frac{1}{2^{(m+n)/2}\Gamma((m+n)/2)}v^{(m+n)/2-1}e^{-v/2}\right)
\end{aligned}
$$

## 020208

以

$$
f_n(x)=\frac{\Gamma({n+1}/{2})}{\sqrt{n\pi}\Gamma({n}/{2})}\left(1+\frac{x^2}{n}\right)^{-(n+1)/2}
$$

表示自由度为 $n$ 的 $t$ 变量的密度函数, 证明:

$$
\lim_{n\to\infty}f_n(x)=\frac{1}{\sqrt{2\pi}}e^{-x^2/2}.
$$​

### 020208-解答

由于

$$
\lim_{n\to\infty}(1+\frac{x^2}{n})^{-\frac{n+1}{2}}\sim e^{-x^2/2},
$$

只需证

$$
\lim_{n\to\infty}\frac{\Gamma({n+1}/{2})}{\Gamma({n}/{2})}\sim\sqrt{n/2}.
$$

由 Stirling 公式和 Gamma 函数,

$$
\Gamma(n)=(n-1)!\sim\sqrt{2\pi(n-1)}\left(\frac{n-1}e\right)^{n-1},
$$

则

$$
\begin{aligned}
&\lim_{n\to\infty}\sqrt{\frac{n-1}{n-2}}\sqrt{n-1}\frac{1}{\sqrt2e}\left(\frac{n-1}{n-2}\right)^{(n-2)/2}\\
&=1\sqrt n\frac{1}{\sqrt2e}e^{1/2}=\sqrt{\frac n2}.
\end{aligned}
$$

## 020210

设随机变量 $X\sim F(1,1)$, 求 $P\{|X|<1\}$​.

### 020210-解答

若 $X\sim F(1,1)$, 则 $1/X\sim F(1,1)$,

则

$$
P\{|X|<1\}=P\left\{\frac{1}{|X|}>1\right\}=P\{|X|>1\},
$$

因此 $P\{|X|<1\}=1/2$.
