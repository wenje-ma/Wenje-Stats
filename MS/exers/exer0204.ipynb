## 020403

设 $X_1,X_2,\cdots,X_n$ 是来自总体 $X$ 的样本, 且 $X$ 的密度函数为

$$
f(x)=
\begin{cases}
\theta x^{\theta-1},&0<x<1,\\
0,&\text{Others},
\end{cases}
$$

其中 $\theta>0$. 证明样本分布族是指数族, 并表示成指数族的自然形式, 求出自然参数空间.

### 020403-解答

由于 $X_1,X_2,\cdots,X_n$ 独立同分布, 则

$$
\begin{aligned}
&f(x_1,\cdots,x_n;\theta)=\theta^n\prod_{i=1}^nx_i^{\theta-1}I_{\{0\le x_i\le1\}}\\
&=\theta^n\exp\left\{\sum_{i=1}^n(\theta-1)\ln x_i\right\}\prod_{i=1}^nI_{\{0\le x_i\le1\}}
\end{aligned}
$$

令 $\varphi=\theta-1$, 则

$$
f(x_1,\cdots,x_n;\varphi)=(\varphi+1)^n\exp\left\{\sum_{i=1}^n\varphi\ln x_i\right\}\prod_{i=1}^nI_{\{0\le x_i\le1\}}.
$$

自然参数空间:

$$
\begin{aligned}
\Theta^*&=\left\{\varphi:\int_{(0,1)^n}\exp\left\{\varphi\sum_{i=1}^n\ln x_i\right\}dx_i<+\infty\right\}\\
&=\left\{\varphi:\prod_{i=1}^n\int_0^1 x_i^{\varphi}dx_i<+\infty\right\}=\{\varphi>-1\}.
\end{aligned}
$$

## 020404

设 $X_1,X_2,\cdots,X_n$ 是来自 Laplace 分布的样本, 其密度函数为

$$
f(x)=\frac{1}{2\theta}e^{-|x|/\theta},
$$

其中 $x\in\mathbb{R}$, $\theta>0$. 证明样本分布族是指数族, 并表示成指数族的自然形式, 求出自然参数空间.

### 020404-解答

由于 $X_1,X_2,\cdots,X_n$ 独立同分布, 则

$$
\begin{aligned}
&f(x_1,\cdots,x_n;\theta)=\frac1{(2\theta)^n}\prod_{i=1}^ne^{-|x_i|/\theta}\\
&=\frac1{(2\theta)^n}\exp\left\{\sum_{i=1}^n-\frac1\theta|x_i|\right\}
\end{aligned}
$$

令 $\varphi=-1/\theta$, 则

$$
f(x_1,\cdots,x_n;\varphi)=\left(-\frac{\varphi}{2}\right)^n\exp\left\{\sum_{i=1}^n\varphi|x_i|\right\}.
$$

自然参数空间:

$$
\begin{aligned}
\Theta^*&=\left\{\varphi:\int_{\mathbb R^n}\exp\left\{\sum_{i=1}^n\varphi |x_i|\right\}dx_i<+\infty\right\}\\
&=\left\{\varphi:2^n\int_{(0,+\infty)^n}\exp\left\{\sum_{i=1}^n\varphi x_i\right\}dx_i<+\infty\right\}\\
&=\left\{\varphi:2^n\prod_{i=1}^n\int_0^{+\infty}e^{\varphi x_i}dx_i<+\infty\right\}=\{\varphi<0\}.
\end{aligned}
$$

## 020405

设 $X_1,X_2,\cdots,X_n$ 是来自 Rayleigh 分布的样本,其密度函数为

$$
f(x)=
\begin{cases}
2\theta xe^{-\theta x^2},&x>0,\\
0,&\text{Others},
\end{cases}
$$

其中 $\theta>0$. 证明样本分布族是指数族, 并表示成指数族的自然形式, 求出自然参数空间.

### 020405-解答

由于 $X_1,X_2,\cdots,X_n$ 独立同分布, 则

$$
f(x_1,x_2,\cdots,x_n;\theta)=(2\theta)^n\exp\left\{\sum_{i=1}^n(-\theta)x_i^2\right\}\prod_{i=1}^nx_iI_{\{x_i>0\}}
$$

令 $\varphi=-\theta$, 则

$$
f(x_1,x_2,\cdots,x_n;\varphi)=(-2\varphi)^n\exp\left\{\sum_{i=1}^n\varphi x_i^2\right\}\prod_{i=1}^nx_iI_{\{x_i>0\}}.
$$

自然参数空间:

$$
\begin{aligned}
\Theta^*&=\left\{\varphi:\int_{(0,+\infty)^n}\exp\left\{\sum_{i=1}^n\varphi x_i^2\right\}\prod_{i=1}^nx_idx_i<+\infty\right\}\\
&=\left\{\varphi:\prod_{i=1}^n\int_0^{+\infty}e^{\varphi x_i^2}x_idx_i<+\infty\right\}=\{\varphi<0\}.
\end{aligned}
$$

## 020406

$$
f(x)=
\begin{cases}

$$
f(x_1,x_2,\cdots,x_n;\lambda)=\lambda^n\exp\left\{\sum_{i=1}^n(-\lambda)x_i\right\}\prod_{i=1}^nI_{\{x_i>0\}}.
$$

令 $\varphi=-\lambda$, 则

$$
f(x_1,x_2,\cdots,x_n;\varphi)=(-\varphi)^n\exp\left\{\sum_{i=1}^n\varphi x_i\right\}\prod_{i=1}^nI_{\{x_i>0\}}.
$$

自然参数空间:

$$
\begin{aligned}
\Theta^*&=\left\{\varphi:\int_{(0,+\infty)^n}\exp\left\{\sum_{i=1}^n\varphi x_i\right\}dx_i<+\infty\right\}\\
&=\left\{\varphi:\prod_{i=1}^n\int_0^{+\infty}e^{\varphi x_i}dx_i<+\infty\right\}=\{\varphi<0\}.
\end{aligned}
$$

## 020407

设 $X_1,X_2,\cdots,X_n$ 是来自对数正态分布的样本, 其密度函数为

$$
f(x;\mu,\sigma^2)=
\begin{cases}
\exp\{-(\ln x-\mu)^2/(2\sigma^2)\}/\sqrt{2\pi}\sigma x,&x>0,\\
0,&\text{Others},
\end{cases}
$$

其中 $\mu\in\mathbb{R}$, $\sigma>0$. 证明样本分布族是指数族, 并表示成指数族的自然形式, 求出自然参数空间.

### 020407-解答

由于 $X_1,X_2,\cdots,X_n$ 独立同分布, 则

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\mu,\sigma^2)\\
&=\left(\frac{1}{\sqrt{2\pi}\sigma}\right)^n\exp\left\{-\frac{1}{2\sigma^2}\sum_{i=1}^n(\ln x_i)^2+\frac{\mu}{\sigma^2}\sum_{i=1}^n\ln x_i-\frac{n\mu^2}{2\sigma^2}\right\}\prod_{i=1}^n\frac1{x_i}I_{\{x_i>0\}}\\
&=\left(\frac{1}{\sqrt{2\pi}\sigma}\right)^n\exp\left\{-\frac{n\mu^2}{2\sigma^2}\right\}\exp\left\{\sum_{i=1}^n\left(-\frac{1}{2\sigma^2}\right)\ln^2x_i+\sum_{i=1}^n\left(\frac{\mu}{\sigma^2}\right)\ln x_i\right\}\prod_{i=1}^n\frac1{x_i}I_{\{x_i>0\}},
\end{aligned}
$$

令 $\varphi_1=-1/2\sigma^2,\varphi_2=\mu/\sigma^2$, 则

$$
\begin{aligned}
&f(x_1,x_2,\cdots,x_n;\mu,\sigma^2)\\
&=\left(-\frac{\varphi_1}{\pi}\right)^{n/2}\exp\left\{\frac{n\varphi_2^2}{4\varphi_1}\right\}\exp\left\{\sum_{i=1}^n\varphi_1\ln^2x_i+\sum_{i=1}^n\varphi_2\ln x_i\right\}\prod_{i=1}^n\frac1{x_i}I_{\{x_i>0\}}
\end{aligned}
$$

自然参数空间:

$$
\begin{aligned}
\Theta^*&=\{(\varphi_1,\varphi_2):\int_{(0,+\infty)^n}\exp\left\{\sum_{i=1}^n\varphi_1\ln^2x_i+\sum_{i=1}^n\varphi_2\ln x_i\right\}\prod_{i=1}^n\frac1{x_i}dx_i<+\infty\}\\
&=\{(\varphi_1,\varphi_2):\prod_{i=1}^n\int_{0}^{+\infty}\exp\{\varphi_1\ln^2x_i+\varphi_2\ln x_i\}\frac1{x_i}dx_i<+\infty\}\\
&=\{(\varphi_1,\varphi_2):\prod_{i=1}^n\exp\left\{-\frac{\varphi_2^2}{4\varphi_1}\right\}\int_{0}^{+\infty}\exp\left\{\varphi_1(\ln x_i+\frac{\varphi_2}{2\varphi_1})^2\right\}\frac1{x_i}dx_i<+\infty\}\\
&=\{(\varphi_1,\varphi_2):\varphi_1<0, \varphi_2\in\mathbb R\}.
\end{aligned}
$$
