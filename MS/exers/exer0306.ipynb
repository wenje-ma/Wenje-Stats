## 030601

设总体 $X$ 的密度函数或分布列为 $f(x,\theta)$, 且 Fisher信息量 $I(\theta)$ 存在, 如果二阶导数 $\partial^2f(x,\theta)/\partial\theta^2$ 对一切 $\theta\in\Theta$ 存在, 证明 Fisher 信息量为

$$
I(\theta)=-E\left(\frac{\partial^2\ln f(X,\theta)}{\partial\theta^2}\right).
$$

### 030601-解答

已知概率密度积分为 $1$,

$$
\int_{-\infty}^{+\infty}f(x;\theta)dx=1,
$$

两侧对 $\theta$ 求偏导,

$$
\int_{-\infty}^{+\infty}\frac{\partial f(x;\theta)}{\partial\theta}dx=0.
$$

对于 Fisher 信息量而言, 想度量的是平均信息, 因此要构造期望的形式:

$$
\frac{\partial f}{\partial\theta}=f\left(\frac1f\frac{\partial f}{\partial\theta}\right)=f\frac{\partial\ln f}{\partial\theta}.
$$

因此

$$
\int_{-\infty}^{+\infty}f\frac{\partial\ln f}{\partial\theta}dx=E\left(\frac{\partial\ln f(X,\theta)}{\partial\theta}\right)=0.
$$

一个自然的想法是继续求导直到结果不为 $0$: 在一阶期望式两侧关于 $\theta$ 继续求导:

$$
\begin{aligned}
&\frac{\partial}{\partial\theta}\int_{-\infty}^{+\infty}\frac{\partial\ln f(X,\theta)}{\partial\theta}fdx\\
&=\int_{-\infty}^{+\infty}\frac{\partial^2\ln f(X,\theta)}{\partial\theta^2}fdx+\int_{-\infty}^{+\infty}\frac{\partial\ln f(X,\theta)}{\partial\theta}\frac{\partial f(X,\theta)}{\partial\theta}dx\\
&=E\left(\frac{\partial^2\ln f(X,\theta)}{\partial\theta^2}\right)+E\left(\frac{\partial\ln f(X,\theta)}{\partial\theta}\right)^2=0.
\end{aligned}
$$

移项得:

$$
E\left(\frac{\partial\ln f(X,\theta)}{\partial\theta}\right)^2=-E\left(\frac{\partial^2\ln f(X,\theta)}{\partial\theta^2}\right).
$$

我们得到了 2 个等价的能表示包含参数信息平均水平的信息量.

## 030602

设总体 $X$ 的分布列为

$$
P\{X=x\}=(x-1)\theta^2(1-\theta)^{x-2},
$$

其中 $0<\theta<1$, $x=2,3,\cdots$, 求 Fisher 信息量 $I(\theta)$.

### 030602-解答

$$
\ln P=\ln(x-1)+2\ln\theta+(x-2)\ln(1-\theta),
$$

$$
\frac{\partial\ln P}{\partial\theta}=\frac{2}{\theta}-\frac{x-2}{1-\theta},
$$

$$
\frac{\partial^2\ln P}{\partial\theta^2}=-\frac{2}{\theta^2}-\frac{x-2}{(1-\theta)^2},
$$

$$
\begin{aligned}
I(\theta)&=-E\left(-\frac{2}{\theta^2}-\frac{X-2}{(1-\theta)^2}\right)\\
&=\frac{2}{\theta^2}+\frac{EX-2}{(1-\theta)^2},
\end{aligned}
$$

其中

$$
\begin{aligned}
EX&=\sum_{x=2}^\infty x(x-1)\theta^2(1-\theta)^{x-2}\\
&=\theta^2\frac{d^2(\sum_{x=2}^\infty(1-\theta)^x)}{d(1-\theta)^2}\\
&=\theta^2\frac{d^2(1-\theta)^2/\theta)}{d(1-\theta)^2}\\
&=\theta^2\frac2{\theta^3}=\frac{2}{\theta}.
\end{aligned}
$$

则

$$
I(\theta)=\frac{2}{\theta^2}+\frac{2/\theta-2}{(1-\theta)^2}=\frac{2}{\theta^2(1-\theta)}.
$$

## 030604

设 $X_1,X_2,\cdots,X_n$ 是来自总体 $X$ 的样本, 且 $X$ 的密度函数为

$$
f(x,\theta)=
\begin{cases}
\exp\{-x/\theta\}/\theta,&x>0,\\
0,&x\le0,
\end{cases}
$$

其中 $\theta>0$ 为未知参数, 求 $\theta$ 的一致最小方差无偏估计, 并证明其方差等于 C-R 下界, 即它是有效估计.

### 030604-解答

已知指数分布是指数族, 且 $EX=\theta$, 故 $\overline{X}$ 即是 $\theta$ 的无偏估计, 另外, $DX=\theta^2$.

$$
\ln f=-\ln\theta-\frac{x}{\theta},
$$

则

$$
\begin{aligned}
I(\theta)&=-E\left(\frac{\partial^2\ln f}{\partial\theta^2}\right)\\
&=-\frac{1}{\theta^2}+\frac{2EX}{\theta^3}=\frac{1}{\theta^2}.
\end{aligned}
$$

因此由 C-R 不等式, 其下界为

$$
\frac{1}{nI(\theta)}=\frac{\theta^2}{n}.
$$

$D\overline{X}=DX/n=\theta^2/n$ 恰为下界, 故是有效估计.

## 030606

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,1)$ 的样本, 其中 $\mu\in\mathbb{R}$ 是未知参数, 证明:

(1)

$\overline{X}^2-1/n$ 是 $\mu^2$ 的 UMVUE;

(2)

$\overline{X}^2-1/n$ 的方差大于 C-R 下界, 即它不是有效估计.

### 030606-解答

(1)

已知 $\overline X$ 充分完备, 因此只需证其无偏性:

$$
\begin{aligned}
&E\left(\overline{X}^2-\frac{1}{n}\right)\\
&=D(\overline{X})+(E\overline{X})^2-\frac{1}{n}\\
&=\frac{1}{n}+\mu^2-\frac{1}{n}=\mu^2.
\end{aligned}
$$

(2)

已知 $f(x,\mu)=\frac{1}{\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2}}$,

$$
\begin{aligned}
I(\theta)&=-E\left(\frac{\partial^2\ln f}{\partial\mu^2}\right)\\
&=-E\left(\partial^2\ln\left(\frac{1}{\sqrt{2\pi}}e^{-(x-\mu)^2/2}\right)\middle/\partial\mu^2\right)=1.
\end{aligned}
$$

由 C-R 不等式, 其下界为

$$
\frac{(g'(\mu))^2}{nI(\theta)}=\frac{(2\mu)^2}{n}=\frac{4\mu^2}{n}.
$$

而

$$
D\overline{X}^2=E\overline{X}^4-(E\overline{X}^2)^2,
$$

对于 $EX^k$, $Z\sim N(0,1)$:

$$
EZ^{2m-1}\equiv0,
$$

因为是奇函数;

$$
\begin{aligned}
EZ^{2m}&=\frac1{\sqrt{2\pi}}\int_{-\infty}^{+\infty}x^{2m}e^{-x^2/2}dx\\
&=\frac{2^m}{\sqrt\pi}\int_0^{+\infty}\left(\frac{x^2}2\right)^{m-1/2}e^{-x^2/2}d\left(\frac{x^2}2\right)\\
&=\frac{2^m}{\sqrt\pi}\Gamma\left(m+\frac12\right).
\end{aligned}
$$

已知 $\Gamma(1/2)=\sqrt\pi$, 则

$$
EZ^2=1,\quad EZ^4=3.
$$

由于 $\overline{X}\sim N(\mu,\frac{1}{n})$, 则

$$
Z=\sqrt n(\overline X-\mu),\quad \overline X=\frac Z{\sqrt n}+\mu.
$$

因此

$$
\begin{aligned}
&D\overline{X}^2=E\overline{X}^4-(E\overline{X}^2)^2\\
&=\frac1{n^2} E(Z^4)+\frac{4\mu^2}n E(Z^2)+\mu^4\\
&+\frac{2\mu^2}n E(Z^2)-\left(\frac1n EZ^2+\mu^2\right)^2\\
&=\frac{2}{n^2}+\frac{4\mu^2}{n}>\frac{4\mu^2}{n},
\end{aligned}
$$

故不是有效估计.
