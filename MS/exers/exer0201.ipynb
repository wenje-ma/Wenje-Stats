## 020102

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体

$$
N(\mu,\sigma^2)
$$

的样本, 其中 $\mu\in\mathbb{R}$, $\sigma^2>0$. 求

$$
T=\sum_{i=1}^nX_i
$$

的抽样分布.

### 020102-解答

由于 $X_i\sim N(\mu_i,\sigma_i^2)$ 且正态分布具有可加性, 则

$$
T=\sum_{i=1}^nX_i\sim N(\sum_{i=1}^n\mu_i,\sum_{i=1}^n\sigma_i^2),
$$

由于同分布的性质 $X_1=X_2=\cdots=X_n=X$, 则

$$
X\sim N(\mu,\sigma^2),
$$

因此

$$
T\sim N(n\mu, n\sigma^2).
$$

## 020103

设 $X\sim B(n,p),0<p<1$, 对 $k\ge1$, 证明:

$$
\begin{aligned}
P\{X\ge r\}&=\sum_{k=r}^n\binom{n}{k}p^k(1-p)^{n-k}\\
&=r\binom{n}{r}\int_0^pt^{r-1}(1-t)^{n-r}dt.
\end{aligned}
$$

### 020103-解答

先证:

$$
P\{X\ge r\}=\sum_{k=r}^n\binom{n}{k}p^k(1-p)^{n-k}.
$$

由于 $X\sim B(n,p),0<p<1$,

则

$$
P\{X=k\}=\binom{n}{k}p^k(1-p)^{n-k},
$$

其中 $k=0,1,\cdots,n$, 这意味着

$$
P\{X\ge r\}=\sum_{k=r}^n\binom{n}{k}p^k(1-p)^{n-k}.
$$

再证:

$$
\sum_{k=r}^n\binom{n}{k}p^k(1-p)^{n-k}=r\binom{n}{r}\int_0^pt^{r-1}(1-t)^{n-r}dt.
$$

记

$$
I(p)=\binom{n}{r}\int_0^pt^{r-1}(1-t)^{n-r}dt,
$$

则只需证

$$
\sum_{k=r}^n\binom{n}{k}p^k(1-p)^{n-k}=rI(p).
$$

两侧对 $p$ 求导, 即只需证

$$
\begin{aligned}
\frac{d}{dp}P\{X\ge r\}&=\sum_{k=r}^n\binom{n}{k}(kp^{k-1}(1-p)^{n-k}\\
&-(n-k)p^k(1-p)^{n-k-1})\\
&=\sum_{k=r}^n\binom{n}{k}kp^{k-1}(1-p)^{n-k}\\
&-\sum_{k=r}^n\binom{n}{k}(n-k)p^k(1-p)^{n-k-1}\\
&=rI'(p)=r\binom{n}{r}p^{r-1}(1-p)^{n-r}.
\end{aligned}
$$

由于组合恒等式, 哑变量换元, 则

$$
\begin{aligned}
&\sum_{k=r}^n\binom{n}{k}kp^{k-1}(1-p)^{n-k}\\
&=\sum_{k=r}^nn\binom{n-1}{k-1}p^{k-1}(1-p)^{n-k}\\
&=\sum_{k=r-1}^{n-1}n\binom{n-1}{k}p^{k}(1-p)^{n-k-1},
\end{aligned}
$$

由于组合恒等式, 则

$$
\begin{aligned}
&\sum_{k=r}^n\binom{n}{k}(n-k)p^k(1-p)^{n-k-1}\\
&=\sum_{k=r}^nn\binom{n-1}{k}p^k(1-p)^{n-k-1},
\end{aligned}
$$

两式相减, 由于 $\binom{n-1}{n}=0$, 组合恒等式, 则

$$
\begin{aligned}
&\frac{d}{dp}P\{X\ge r\}\\
&=n\binom{n-1}{r-1}p^{r-1}(1-p)^{n-r}\\
&-n\binom{n-1}{n}p^{r-1}(1-p)^{n-r}\\
&=r\binom{n}{r}p^{r-1}(1-p)^{n-r}=rI'(p).
\end{aligned}
$$

依据 Newton-Leibniz 公式:

$$
\begin{aligned}
P\{X\ge r\}&=r\int_0^pI'(t)dt=rI(p)+C.
\end{aligned}
$$

取 $p=0$, 即发生事件的概率为 $0$, 这意味着

$$
P\{X\ge r\}|_{p=0}=0=0+C=rI(0)+C,
$$

即 $C=0$, 因此

$$
P\{X\ge r\}=rI(p).
$$

## 020105

设 $X_1,X_2,\cdots,X_n$ 是来自指数分布总体 $E(\lambda)$ 的样本,其密度函数为

$$
f(x)=
\begin{cases}
\lambda e^{-\lambda x},&x>0,\\
0,&x\le0,
\end{cases}
$$

其中 $\lambda>0$. 求:

(1)

$X_{(1)}$ 的密度函数;

(2)

求解 $E(X_{(1)})$, $D(X_{(1)})$.

### 020105-解答

(1)

依据 Newton-Leibniz 公式:

$$
\begin{aligned}
F(x)&=\int_{-\infty}^xf(x)dx\\
&=\int_{-\infty}^00dx+\int_0^xf(x)dx\\
&=1-e^{-\lambda x},
\end{aligned}
$$

则由分布函数的定义, 顺序统计量的定义, 独立的定义, 同分布的性质 $X_1=X_2=\cdots=X_n=X$, 分布函数的定义得

$$
\begin{aligned}
F_{X_{(1)}}(x)&=P\{X_{(1)}\le x\}\\
&=1-P\{X_1>x, X_2>x, s, X_n>x\}\\
&=1-(P\{X>x\})^n\\
&=1-(1-F(x))^n\\
&=1-e^{-n\lambda x},\quad x>0.
\end{aligned}
$$

因此

$$
f_{X_{(1)}}(x)=F_{(1)}'(x)=n\lambda e^{-n\lambda x}I_{\{x>0\}}.
$$

(2)

由 (1) 中求得的密度函数 $f_{X_{(1)}}(x)$ 知

$$
X_{(1)}\sim E(n\lambda),
$$

则

$$
E(X_{(1)})=\frac{1}{n\lambda},
$$

$$
D(X_{(1)})=\frac{1}{n^2\lambda^2}.
$$

## 020106

设 $X_1,X_2,\cdots,X_9$ 是来自正态总体 $N(1,9)$ 的样本, 求:

(1)

$$
P\{X_{(1)}>4\}.
$$

(2)

$$
P\{X_{(9)}<4\}.
$$

### 020106-解答

(1)

由顺序统计量的定义, 独立的定义, 同分布的性质 $X_1=X_2=\cdots=X_9=X$, 分布函数的定义可得

$$
\begin{aligned}
&P\{X_{(1)}>4\}\\
&=P\{X_1>4,X_2>4,\cdots,X_9>4\}\\
&=\prod_{i=1}^9P\{X_i>4\}\\
&=P\{X>4\}^9=(1-F(X\le4))^9.
\end{aligned}
$$

由于 $X_i\sim N(1,9)$, 因此将其标准化:

$$
F(X\le4)=F(\frac{X-1}3\le\frac{4-1}3)=\Phi(1),
$$

查表得 $\Phi(1)=0.8413$, 则 $F(X\le4)=\Phi(1)=0.8413$,

$$
P\{X_{(1)}>4\}=(1-0.8413)^9.
$$

(2)

由顺序统计量的定义, 独立的定义, 同分布的性质 $X_1=X_2=\cdots=X_9=X$, 分布函数的定义可得

$$
\begin{aligned}
&P\{X_{(9)}<4\}\\
&=P\{X_1<4,X_2<4,\cdots,X_9<4\}\\
&=\prod_{i=1}^9P\{X_i<4\}\\
&=P\{X<4\}^9=F(X\le4)^9=(0.8413)^9.
\end{aligned}
$$

## 020107

设 $X_1,X_2,\cdots,X_n$ 是来自均匀分布总体

$$
U[\theta_1,\theta_2],
$$

的样本, 其中 $\theta_1<\theta_2$. 分别求顺序统计量 $X_{(1)}$, $X_{(n)}$ 的密度函数.

### 020107-解答

由于 $X\sim U[\theta_1,\theta_2]$, 则

$$
F(x)=
\begin{cases}
0,&x<\theta_1,\\
(x-\theta_1)/(\theta_{2}-\theta_1),&\theta_1\le x\le\theta_2,\\
1,&x>\theta_2,
\end{cases}
$$

因此

$$
f(x)=\frac1{\theta_2-\theta_1}I_{\{\theta_1\le x\le\theta_2\}}.
$$

由顺序统计量的定义, 独立的定义, 同分布的性质 $X_1=X_2=\cdots=X_n=X$, 分布函数的定义可得

$$
\begin{aligned}
&P\{X_{(1)}<x\}\\
&=1-P\{X_{(1)}\ge x\}\\
&=1-P\{X_1\ge x,X_2\ge x,\cdots,X_n\ge x\}\\
&=1-\prod_{i=1}^nP\{X_i\ge x\}\\
&=1-P\{X\ge x\}^n\\
&=1-(1-P\{X\le x\})^n\\
&=1-(1-F(x))^n,
\end{aligned}
$$

因此

$$
\begin{aligned}
f_{X_{(1)}}(x)&=(1-(1-F(x))^n)'\\
&=n(1-F(x))^{n-1}f(x)\\
&=\frac{n(\theta_2-x)^{n-1}}{(\theta_2-\theta_1)^n}I_{\{\theta_1\le x\le\theta_2\}}.
\end{aligned}
$$

由顺序统计量的定义, 独立的定义, 同分布的性质 $X_1=X_2=\cdots=X_n=X$, 分布函数的定义可得

$$
\begin{aligned}
&P\{X_{(n)}<x\}\\
&=P\{X_1\le x,X_2\le x,\cdots,X_n\le x\}\\
&=\prod_{i=1}^nP\{X_i\le x\}\\
&=P\{X\le x\}^n=F(x)^n,
\end{aligned}
$$

因此

$$
\begin{aligned}
f_{X_{(n)}}(x)&=(F(x)^n)'\\
&=nF^{n-1}(x)f(x)\\
&=\frac{n(x-\theta_1)^{n-1}}{(\theta_2-\theta_1)^n}I_{\{\theta_1\le x\le\theta_2\}}.
\end{aligned}
$$
