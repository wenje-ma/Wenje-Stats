## 050301

某公司生产一种 $40\,\mathrm{W}$ 灯泡, 并声称灯泡的平均寿命为 $1000\,\mathrm{h}$. 消费者随机地抽取了4个灯泡做试验, 得到它们的寿命分别是

$$
980\,\mathrm{h},\quad1005\,\mathrm{h},\quad992\,\mathrm{h},\quad973\,\mathrm{h},
$$

计算得平均寿命为 $987.5\,\mathrm{h}$. 若假定灯泡的寿命服从指数分布, 问在显著性水平 $0.05$ 下, 有无充分证据认为平均寿命小于 $1000\,\mathrm{h}$?

### 050301-解答

提出假设:

$$
H_0:\mu=1/\lambda\ge1000;\quad H_1:\mu=1/\lambda<1000.
$$

构造检验统计量:

$$
Q=2n\lambda\overline{X}\sim\chi^2(2n).
$$

拒绝域 $D=\{Q\le\chi^2_{1-\alpha}(2n)\}$, 计算

$$
Q=7.9>2.733=\chi^2_{0.95}(8).
$$

无法拒绝 $H_0$, 即无充分证据认为寿命小于 $1000\,\mathrm{h}$.

## 050303

设 $X_1,X_2,\cdots,X_m$ 是来自指数分布总体 $E(\lambda_1)$ 的样本, 其中 $\lambda_1>0$ 为未知参数, $Y_1,Y_2,\cdots,Y_n$ 是来自指数分布总体 $E(\lambda_2)$ 的样本, 其中 $\lambda_2>0$ 为未知参数, 且两样本相互独立, 在显著性水平 $\alpha$ 下, 检验假设

$$
H_0:\lambda_1=\lambda_2;\quad H_1:\lambda_1\ne\lambda_2.
$$

### 050303-解答

构造检验统计量:

$$
Q_1=2m\lambda_1\overline{X}\sim\chi^2(2m),
$$

$$
Q_2=2n\lambda_2\overline{Y}\sim\chi^2(2n),
$$

$$
F=\frac{Q_1/2m}{Q_2/2n}=\frac{\overline{X}}{\overline{Y}}\sim F(2m,2n).
$$

则拒绝域

$$
D=\left\{\frac{\overline X}{\overline Y}\le F_{1-\alpha/2}(2m,2n)\,\text{or}\,\frac{\overline X}{\overline Y}\ge F_{\alpha/2}(2m,2n)\right\}.
$$

## 050304

设 $X_1,X_2,\cdots,X_n$ 为来自总体 $X$ 的样本, 且 $X\sim P(\lambda)$, 其中 $\lambda>0$ 为未知参数, 在显著性水平 $\alpha$ 下, 用大样本方法检验假设

$$
H_0:\lambda=\lambda_0;\quad H_1:\lambda\ne\lambda_0.
$$

### 050304-解答

由 CLT, 大样本下

$$
\lambda=\overline{X}\sim N\left(\lambda,\frac{\lambda}{n}\right).
$$

构造检验统计量:

$$
Z=\frac{\overline{X}-\lambda_0}{\sqrt{\lambda_0/n}}\sim N(0,1).
$$

拒绝域

$$
D=\left\{\left|\frac{\overline{X}-\lambda_0}{\sqrt{\lambda_0/n}}\right|\ge u_{\alpha/2}\right\}.
$$

## 050306

某市随机抽取 1000 个家庭, 调查其中有 450 家拥有电脑, $p$ 表示该市拥有电脑家庭的比例, 在显著性水平 $\alpha=0.05$ 下, 检验假设

$$
H_0:p=0.5;\quad H_1:p\ne0.5.
$$

### 050306-解答

拥有电脑的家庭数 $X\sim B(n,p)$,

$$
E(X)=np,\quad DX=np(1-p).
$$

则拥有电脑的比例 $\widehat{p}=X/n$,

$$
E(\widehat{p})=p,\quad D(\widehat{p})=\frac{p(1-p)}{n}.
$$

在原假设 $H_0$ 成立下, $p=0.5$, 构造检验统计量:

$$
Z=\frac{\widehat{p}-p}{\sqrt{p(1-p)/n}}
  =\frac{0.45-0.5}{\sqrt{0.5\times0.5/1000}}
  =-3.162,
$$

拒绝域 $D=\{|Z|\ge u_{0.025}\}$, 计算得

$$
3.162>1.96,
$$

因此拒绝 $H_0$, 即该市拥有电脑的比例并非 $50%$.
