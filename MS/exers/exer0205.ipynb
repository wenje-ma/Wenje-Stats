## 020501

设 $X_1,X_2,\cdots,X_n$ 是来自 Poisson 分布总体 $P(\lambda)$, $\lambda >0$ 的样本,

(1)

分别用定义法和因子分解定理法证明

$$
\sum_{i=1}^nX_i
$$

为充分统计量;

(2)

若 $n=2$, 则 $X_1+2X_2$ 不是充分统计量.

### 020501-解答

(1)

由于 $X_1,X_2,\cdots,X_n$ 独立同分布, 则

$$
P(X_1=x_1,X_2=x_2,\cdots,X_n=x_n)=\frac{\lambda^{\sum_{i=1}^nx_i}e^{-n\lambda}}{\prod_{i=1}^nx_i!}.
$$

- 定义法

令 $T=\sum_{i=1}^nX_i$, 则

$$
\begin{aligned}
&P(X_1=x_1,\cdots,X_n=x_n|T=t)\\
&=\frac{P(X_1=x_1,X_2=x_2,\cdots,X_n=x_n,T=t)}{P(T=t)}\\
&=\frac{P(X_1=x_1,X_2=x_2,\cdots,X_n=x_n)}{P(T=t)}.
\end{aligned}
$$

由于 Poisson 分布具有再生性, 则 $T\sim P(n\lambda)$,

$$
P(T=t)=\frac{(n\lambda)^te^{-n\lambda}}{t!},
$$

因此

$$
P(X_1=x_1,X_2=x_2,\cdots,X_n=x_n|T=t)=\frac{t!}{n^t\prod_{i=1}^nx_i!}.
$$

与 $\lambda$ 无关, 故为充分统计量.

- 因子分解定理

$$
P(X_1=x_1,X_2=x_2,\cdots,X_n=x_n;\lambda)=\frac{\lambda^{\sum_{i=1}^nx_i}e^{-n\lambda}}{\prod_{i=1}^nx_i!}.
$$

因此

$$
T(X_1,X_2s,X_n)=\sum_{i=1}^nX_i
$$

为充分统计量.

(2)

若 $n=2$, 则

$$
P\{X_1=x_1,X_2=x_2\}=\frac{\lambda^{x_1+x_2}e^{-2\lambda}}{x_1^2x_2}.
$$

反证: 若 $X_1+2X_2$ 是充分统计量, 则

$$
f(x_1,x_2;\theta)=\frac{\lambda^{x_1+2x_2}e^{-2\lambda}}{x_1^2x_2\lambda^{x_2}},
$$

此时 $h(x_1,x_2)=1/x_1^2x_2\lambda^{x_2}$ 中含 $\lambda$, 矛盾.

## 020505

设 $X_1,X_2,\cdots,X_n$ 是来自 Rayleigh 分布的样本,其密度函数为

$$
f(x)=
\begin{cases}
2\theta xe^{-\theta x^2},&x>0,\\
0,&\text{Others},
\end{cases}
$$

其中 $\theta>0$. 试给出一个充分统计量.

### 020505-解答

由于 $X_1,X_2,\cdots,X_n$ 独立同分布, 因子分解定理:

$$
\begin{aligned}
f(x_1,\cdots,x_n;\theta)&=(2\theta)^n\prod_{i=1}^nx_ie^{-\theta\sum_{i=1}^nx_i^2}I_{\{x_i>0\}}\\
&=(2\theta)^ne^{-\theta\sum_{i=1}^nx_i^2}\prod_{i=1}^nx_iI_{\{x_i>0\}}
\end{aligned}
$$

因此

$$
T(X_1,X_2,\cdots,X_n)=\sum_{i=1}^nX_i^2.
$$

为充分统计量.

## 020506

设 $X_1,X_2,\cdots,X_n$ 是来自指数分布 $E(\lambda)$ 的样本, 其密度函数为

$$
f(x)=
\begin{cases}
\lambda e^{-\lambda x},&x>0,\\
0,&x\le0,
\end{cases}
$$

其中 $\lambda>0$. 证明 $\sum_{i=1}^nX_i$ 是充分统计量.

### 020506-解答

由于 $X_1,X_2,\cdots,X_n$ 独立同分布, 因子分解定理:

$$
f(x_1,\cdots,x_n;\lambda)=\lambda^ne^{-\lambda\sum_{i=1}^nx_i}\prod_{i=1}^nI_{\{x_i>0\}},
$$

因此

$$
T(X_1,X_2,\cdots,X_n)=\sum_{i=1}^nX_i.
$$

为充分统计量.

## 020507

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本.

(1)

在 $\mu$ 已知时, 给出 $\sigma^2$ 的一个充分统计量;

(2)

在 $\sigma^2$ 已知时, 给出 $\mu$ 的充分统计量.

### 020507-解答

由于 $X_1,X_2,\cdots,X_n$ 独立同分布, 因子分解定理:

$$
f(x_1,\cdots,x_n;\mu,\sigma^2)=(2\pi\sigma^2)^{-n/2}\exp\left\{-\frac{1}{2\sigma^2}\sum_{i=1}^n(x_i-\mu)^2\right\}.
$$

(1)

在 $\mu$ 已知时,

$$
T(X_1,X_2,\cdots,X_n)=\sum_{i=1}^n(X_i-\mu)^2.
$$

为充分统计量.

(2)

$$
\begin{aligned}
&f(x_1,\cdots,x_n;\mu,\sigma^2)\\
&=(2\pi\sigma^2)^{-n/2}\exp\left\{-\frac{1}{2\sigma^2}\sum_{i=1}^n(x_i-\mu)^2\right\}\\
&=(2\pi\sigma^2)^{-n/2}\exp\left\{\frac{\mu}{\sigma^2}\sum_{i=1}^nx_i-\frac{n\mu^2}{2\sigma^2}\right\}\exp\left\{-\frac{1}{2\sigma^2}\sum_{i=1}^nx_i^2\right\}.
\end{aligned}
$$

因此在 $\sigma^2$ 已知时,

$$
T(X_1,X_2,\cdots,X_n)=\sum_{i=1}^nX_i.
$$

为充分统计量.
