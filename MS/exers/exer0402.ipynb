## 040203

设轴承内环的锻压零件的高度服从正态分布 $N(\mu,\,\sigma^2)$. 现从中随机抽取 $20$ 只内环,\, 测得高度(单位: mm)为

$$
\begin{aligned}
&32.27,\,32.56,\,32.73,\,32.98,\,31.94,\,32.50,\,32.09,\,\\
&32.33,\,32.43,\,32.13,\,32.72,\,32.41,\,32.05,\,32.19,\,\\
&32.15,\,32.55,\,31.51,\,32.16,\,31.92,\,32.28.
\end{aligned}
$$

求内环平均高度的置信水平分别为 $95\%$ 和 $99\%$ 的置信区间. 

### 040203-解答

$$
\begin{aligned}
\overline{x}&=\frac{1}{20}\sum_{i=1}^{20}x_i=32.29,\\
s^2&=\frac{1}{n-1}(\sum x_i^2-n\overline{x}^2)=0.1355,\\
s&=\sqrt{s^2}=0.3684,
\end{aligned}
$$

$\sigma$ 未知, 求 $\mu$ 的区间估计:

$$
\frac{\overline{X}-\mu}{S/\sqrt{n}}\sim t(n-1),
$$

$$
P\{|T|<t_{\alpha/2}(n-1)\}=1-\alpha,
$$

解得置信区间为: 

$$
\left[\overline{X}-\frac{S}{\sqrt{n}}t_{\alpha/2}(n-1),\overline{X}+\frac{S}{\sqrt{n}}t_{\alpha/2}(n-1)\right].
$$

- $\alpha=0.05$ 时, 区间为 $[32.1390,32.4510]$;
- $\alpha=0.01$ 时, 区间为 $[32.0817, 32.5083]$.

## 040205

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本, $\sigma^2=\sigma_0^2$ 已知, 问样本容量 $n$ 取多大时才能保证 $\mu$ 的置信水平为 $1-\alpha(0<\alpha<1)$ 的置信区间的长度不大于 $k$?

### 040205-解答

$\sigma$ 已知, 求 $n$ 使 $\mu$ 的区间估计长度不大于 $k$: 

$$
\frac{\overline{X}-\mu}{\sigma/\sqrt{n}}\sim N(0,1),
$$

$$
P\{|U|<u_{\alpha/2}\}=1-\alpha.
$$

解得置信区间为: 

$$
\left[\overline{X}-\frac{\sigma}{\sqrt{n}}u_{\alpha/2},\overline{X}+\frac{\sigma}{\sqrt{n}}u_{\alpha/2}\right].
$$

区间长度:

$$
\frac{2\sigma}{\sqrt{n}}u_{\alpha/2}\le k,
$$

解得: 

$$
n\ge\left\lceil(\frac{2\sigma u_{\alpha/2}}{k})^2\right\rceil.
$$

## 040209

为了解一台测量长度的仪器的精度, 对一根长 30mm 的标准金属棒进行了 6 次测量, 结果(单位: mm)是

$$
30.1,\,29.9,\,29.8,\,30.3,\,30.2,\,29.6.
$$

假如测量值服从正态分布 $N(30,\sigma^2)$, 求 $\sigma^2$ 的置信水平为 $95\%$ 的置信区间.

### 040209-解答

$\mu$ 已知, 求 $\sigma^2$ 的区间估计:

$$
S^2=\frac{1}{n}\sum(X_i-\mu)^2,
$$

$$
T=\frac{nS^2}{\sigma^2}\sim\chi^2(n),
$$

$$
P\{C_1\le T\le C_2\}=1-\alpha,
$$

为计算方便, 取

$$
C_1=\chi^2_{1-\alpha/2}(n),C_2=\chi^2_{\alpha/2}(n).
$$

置信区间为:

$$
\left[\frac{nS^2}{\chi^2_{\alpha/2}(n)},\frac{nS^2}{\chi^2_{1-\alpha/2}(n)}\right].
$$

已知 $\mu=30$, 则

$$
\sum_{i=1}^6(X_i-\mu)^2=0.35.
$$

$\alpha=0.05$ 时, 置信区间为 $[0.0242, 0.2829]$.

## 040211

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本, $a,b$ 为常数, 且 $0<a<b$, 求随机区间

$$
\left[\frac{\sum_{i=1}^n(X_i-\overline{X})^2}{b},\frac{\sum_{i=1}^n(X_i-\overline{X})^2}{a}\right]
$$

的长度 $L$ 的数学期望和方差.

### 040211-解答

已知

$$
S^2=\frac{1}{n-1}\sum(X_i-\overline{X})^2,
$$

$$
\frac{(n-1)S^2}{\sigma^2}=\frac{1}{\sigma^2}\sum(X_i-\overline{X})^2\sim\chi^2(n-1),
$$

区间长度

$$
L=\left(\frac{1}{b}-\frac{1}{a}\right)\sum_{i=1}^n(X_i-\overline{X})^2,
$$

则令

$$
U=\frac{L}{(1/b-1/a)\sigma^2}\sim\chi^2(n-1),
$$

$$
\begin{aligned}
E(L)&=\left(\frac{1}{b}-\frac{1}{a}\right)\sigma^2E(U)\\
&=\left(\frac{1}{b}-\frac{1}{a}\right)\sigma^2(n-1),
\end{aligned}
$$

$$
\begin{aligned}
D(L)&=\left(\frac{1}{b}-\frac{1}{a}\right)^2\sigma^4D(U)\\
&=2\left(\frac{1}{b}-\frac{1}{a}\right)^2\sigma^4(n-1).
\end{aligned}
$$

## 040214

随机选取 $9$ 发炮弹, 测得炮弹的炮口速度的样本标准差 $s=11\,\mathrm{m/s}$, 若炮弹的炮口速度服从正态分布 $N(\mu,\sigma^2)$, 其中 $\mu,\sigma^2$ 未知, 求标准差 $\sigma$ 的置信水平为 $95\%$ 的单侧置信上限.

### 040214-解答

$\mu$ 未知, 求 $\sigma$ 的区间估计:

$$
\frac{(n-1)S^2}{\sigma^2}\sim\chi^2(n-1),
$$

$$
P\{T\ge\chi^2_{1-\alpha}(n-1)\}=1-\alpha,
$$

则 $\sigma$ 的单侧置信上限为:

$$
S\sqrt{\frac{n-1}{\chi^2_{1-\alpha}(n-1)}},
$$

代入数据得 $18.82\,\mathrm{m/s}$.

## 040220

两台机床加工同一种零件, 分别抽取 $6$ 个和 $9$ 个零件, 测量其长度并算得样本方差分别为

$$
s_1^2=0.245,\quad s_2^2=0.357.
$$

假定两台机床加工零件的长度分别服从正态分布 $N(\mu_1,\sigma_1^2)$ 和 $N(\mu_2,\sigma_2^2)$. 求两总体标准差之比 $\sigma_1/\sigma_2$ 的置信水平为 $95\%$ 的置信区间.

### 040220-解答

$\mu_1,\mu_2$ 未知, 求 $\sigma_1/\sigma_2$ 的置信区间:

$$
\frac{S_1^2/S_2^2}{\sigma_1^2/\sigma_2^2}\sim F(m-1,n-1),
$$

$$
P\{F_{1-\alpha/2}(m-1,n-1)\le T\le F_{\alpha/2}(m-1,n-1)\}=1-\alpha,
$$

则 $\sigma_1^2/\sigma_2^2$ 的置信区间为:

$$
\left[\frac{1}{F_{\alpha/2}(m-1,n-1)}\frac{S_1^2}{S_2^2},\frac{1}{F_{1-\alpha/2}(m-1,n-1)}\frac{S_1^2}{S_2^2}\right].
$$

开方得 $\sigma_1/\sigma_2$ 的置信区间:

$$
\left[\frac{S_1}{S_2}\frac{1}{\sqrt{F_{\alpha/2}(m-1,n-1)}},\frac{S_1}{S_2}\frac{1}{\sqrt{F_{1-\alpha/2}(m-1,n-1)}}\right].
$$

代入数据得: $[0.3775,2.1534]$.

## 040221

设从总体 $X\sim N(\mu_1,\sigma_1^2)$ 和总体 $Y\sim N(\mu_2,\sigma_2^2)$ 中分别抽取容量为 $m=10,n=15$ 的样本, 计算得样本均值分别为 $\overline{x}=82,\overline{y}=76$, 样本方差分别为 $s_1^2=56.5,s_2^2=52.4$.

(1)

若已知 $\sigma_1^2=64,\sigma_2^2=49$, 求 $\mu_1-\mu_2$ 的置信水平为 $95\%$ 的置信区间;

(2)

若 $\sigma_1^2=\sigma_2^2$ 未知, 求 $\mu_1-\mu_2$ 的置信水平为 $95\%$ 的置信区间;

(4)

求 $\sigma_1^2/\sigma_2^2$ 的置信水平为 $95\%$ 的置信区间.

### 040221-解答

(1)

$\sigma_1^2,\sigma_2^2$ 已知, 求 $\mu_1-\mu_2$ 的置信区间:

$$
\frac{(\overline{Y}-\overline{X})-(\mu_2-\mu_1)}{\sqrt{\sigma_1^2/m+\sigma_2^2/n}}\sim N(0,1),
$$

置信区间为:

$$
\left[\overline{Y}-\overline{X}-u_{\frac{\alpha}{2}}\sqrt{\frac{\sigma_1^2}{m}+\frac{\sigma_2^2}{n}},\overline{Y}-\overline{X}+u_{\frac{\alpha}{2}}\sqrt{\frac{\sigma_1^2}{m}+\frac{\sigma_2^2}{n}}\right],
$$

代入数据得: $[-0.0939,12.0939]$.

(2)

$\sigma_1^2=\sigma_2^2=\sigma^2$ 未知, 求 $\mu_1-\mu_2$ 的置信区间:

$$
S_w^2=\frac{(m-1)S_1^2+(n-1)S_2^2}{m+n-2},
$$

$$
\frac{(\overline{Y}-\overline{X})-(\mu_2-\mu_1)}{S_w\sqrt{1/m+1/n}}\sim t(m+n-2),
$$

置信区间为:

$$
\left[\overline{Y}-\overline{X}-S_wt_{\alpha/2}(m+n-2)\sqrt{\frac{1}{m}+\frac{1}{n}},\overline{Y}-\overline{X}+S_wt_{\alpha/2}(m+n-2)\sqrt{\frac{1}{m}+\frac{1}{n}}\right],
$$

代入数据得: $[-0.2063,12.2063]$.

(4)

$\mu_1,\mu_2$ 未知, 求 $\sigma_1^2/\sigma_2^2$ 的置信区间:

$$
\frac{S_1^2/S_2^2}{\sigma_1^2/\sigma_2^2}\sim F(m-1,n-1),
$$

则 $\sigma_1^2/\sigma_2^2$ 的置信区间为:

$$
\left[\frac{S_1^2}{S_2^2}F_{\alpha/2}(n-1,m-1),\frac{S_1^2}{S_2^2}F_{1-\alpha/2}(n-1,m-1)\right],
$$

代入数据得: $[0.3359,4.0973]$.
