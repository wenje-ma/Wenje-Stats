## 020302

设 $X_1,X_2,\cdots,X_n$ 是来自 Poisson 分布总体 $P(\lambda)$, $\lambda>0$ 的样本, 求

$$
E(\overline{X}),\quad D(\overline{X}),\quad E(S^2).
$$

### 020302-解答

已知 $EX_i=\lambda$, 则

$$
E\overline{X}=\frac{1}{n}\sum_{i=1}^n\lambda=\lambda.
$$

已知 $DX_i=\lambda$, 则

$$
D\overline{X}=\frac{1}{n^2}\sum_{i=1}^n\lambda=\frac{\lambda}{n}.
$$

已知 $S^2=\frac{1}{n-1}(\sum_{i=1}^nX_i^2-n\overline{X}^2)$, 则

$$
ES^2=\frac{1}{n-1}\left(\sum_{i=1}^nEX_i^2-nE\overline{X}^2\right).
$$

由于 $D(X_i)=EX_i^2-(EX_i)^2$, 则

$$
EX_i^2=\lambda^2+\lambda.
$$

由于 $D(\overline{X})=E\overline{X}^2-(E\overline{X})^2$, 则

$$
E\overline{X}^2=\frac{\lambda}{n}+\lambda^2.
$$

因此

$$
ES^2=\frac{1}{n-1}(n\lambda^2+n\lambda-\lambda-n\lambda^2)=\lambda.
$$

## 020304

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$, $\mu\in\mathbb{R},\sigma^2>0$ 的样本, 求

$$
E(\overline{X}),\quad D(\overline{X}),\quad E(S^2),\quad D(S^2).
$$

### 020304-解答

已知 $EX_i=\mu$, 则

$$
E\overline{X}=\frac{1}{n}\sum_{i=1}^n\mu=\mu.
$$

已知 $DX_i=\sigma^2$, 则

$$
D\overline{X}=\frac{1}{n^2}\sum_{i=1}^n\sigma^2=\frac{\sigma^2}{n}.
$$

已知

$$
S^2=\frac{1}{n-1}\left(\sum_{i=1}^nX_i^2-n\overline{X}^2\right),
$$

$$
D(X_i)=EX_i^2-(EX_i)^2,
$$

$$
D(\overline{X})=E\overline{X}^2-(E\overline{X})^2,
$$

因此

$$
ES^2=\frac{1}{n-1}(n\sigma^2+n\mu^2-\sigma^2-n\mu^2)=\sigma^2.
$$

已知 $(n-1)S^2/\sigma^{2}\sim\chi^2(n-1)$, 因此

$$
D\left(\frac{(n-1)S^2}{\sigma^2}\right)=2(n-1),
$$

$$
D(S^2)=\frac{2\sigma^4}{n-1}.
$$

## 020305

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本, $\mu\in\mathbb{R}$, $\sigma^2>0$ 的样本, 且

$$
\overline{X}=\frac{1}{n}\sum_{i=1}^nX_i,\quad S_n^2=\frac{1}{n}\sum_{i=1}^n(X_i-\overline{X})^2,
$$

又设 $X_{n+1}\sim N(\mu,\sigma^2)$, 且 $X_{n+1}$ 与 $X_1,X_2,\cdots,X_n$ 独立, 求常数 $c$ 使得

$$
\frac{c(X_{n+1}-\overline{X})}{S_n}
$$

服从 $t$ 分布, 并指出分布的自由度.

### 020305-解答

由于 $X_i\sim N(\mu,\sigma^2)$ 与 $\overline{X}\sim N(\mu,\frac{\sigma^2}{n})$ 独立, 则正态分布具有可加性, 因此

$$
X_{n+1}-\overline{X}\sim N(0,\frac{n+1}{n}\sigma^2).
$$

由于 $S_n^2=\frac{1}{n}\sum_{i=1}^n(X_i-\overline{X})^2$, 则

$$
\frac{nS_n^2}{\sigma^2}\sim \chi^2(n-1).
$$

因此

$$
\frac{(X_{n+1}-\overline X)/(\sigma\sqrt{(n+1)/n})}{\sqrt{(nS_n^2)/((n-1)\sigma^2)}}\sim t(n-1),
$$

即 $c=\sqrt{(n-1)/(n+1)}$, 自由度为 $n-1$.

## 020307

设 $X_1,X_2,\cdots,X_{m+n}$ 是来自正态总体 $N(0,\sigma^2)$, $\sigma^2>0$ 的样本, 求常数 $c$ 使得

$$
F=\left.c\sum_{i=1}^nX_i^2\middle/\sum_{i=n+1}^{m+n}X_i^2\right.
$$

服从 $F$ 分布,并指出分布的自由度.

### 020307-解答

由于 $\chi^2$ 分布的定义,

$$
\sum_{i=1}^n\left(\frac{X_i}{\sigma}\right)^2\sim\chi^2(n),
$$

$$
\sum_{i=n+1}^{m+n}\left(\frac{X_i}{\sigma}\right)^2\sim\chi^2(m).
$$

由于

$$
\frac{\sum_{i=1}^n(X_i/\sigma)^2/n}{\sum_{i=n+1}^{m+n}(X_i/\sigma)^2/m}\sim F(n,m),
$$

则 $c=m/n$, 第一自由度为 $n$, 第二自由度为 $m$.

## 020308

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$, $\mu\in\mathbb{R}$, $\sigma^2>0$ 的样本, 求常数 $a$ 和 $b$ 使得

$$
a(\overline{X}-\mu)^2+bS^2\sim \chi^2(n)​.
$$

### 020308-解答

由于 $\overline{X}\sim N(\mu,\sigma^2/n)$, $(n-1)S^2/\sigma^2\sim\chi^2(n-1)$, $\chi^2$ 分布的再生性, 有

$$
\left(\frac{\overline X-\mu}{\sigma/\sqrt n}\right)^2+\frac{(n-1)S^2}{\sigma^2}\sim\chi^2(n),
$$

即 $a=n/\sigma^2$, $b=(n-1)/\sigma^2$.

## 020309

设 $X_1,X_2$ 是来自正态总体 $N(0,\sigma^2)$, $\sigma^2>0$ 的样本, 求

$$
Y=\left(\frac{X_1+X_2}{X_1-X_2}\right)^2
$$

的分布, 并写出分布的自由度.

### 020309-解答

由于 $X_1,X_2\sim N(0,\sigma^2)$ 独立, 则正态分布具有可加性, 因此

$$
X_1+X_2\sim N(0,2\sigma^2),
$$

$$
X_1-X_2\sim N(0,2\sigma^2).
$$

即

$$
Y=\frac{((X_1+X_2)/{\sqrt{2}\sigma})^2/1}{((X_1-X_2)/{\sqrt{2}\sigma})^2/1}\sim F(1,1),
$$

即 $Y$ 为 $F$ 分布, 第一自由度和第二自由度都是 $1$.

## 020312

设 $X_1,X_2,X_3,X_4$ 是来自正态总体 $N(0,\sigma^2)$, $\sigma^2>0$ 的样本, 记

$$
Y=\frac{X_1}{\sqrt{X_2^2+X_3^2+X_4^2}}.
$$

(1)

求 $c$ 使得 $cY$ 服从 $t$ 分布, 并指出分布的自由度;

(2)

求 $E(cY)$, $D(cY)$​.

### 020312-解答

(1)

由于 $X_i\sim N(0,\sigma)$, $(X_2^2+X_3^2+X_4^2)/\sigma^2\sim \chi^2(3)$,

则标准化得:

$$
\frac{X_1}{\sigma}\sim N(0,1).
$$

由于

$$
\frac{X_1/\sigma}{\sqrt{X_2^2+X_3^2+X_4^2/3\sigma^2}}\sim t(3),
$$

由于 $t$ 分布的对称性, $c=\pm\sqrt3$, 分布的自由度为 3.

(2)

由于 $ET=0$, $n>1$, $DT=\frac{n}{n-2}$, $n>2$, 则

$$
E(cY)=0,\quad D(cY)=3.
$$

## 020319

设 $X_1,X_2,\cdots,X_n$ 是来自均匀分布总体 $U(0,\theta)$, $\theta>0$ 的样本, 求

$$
\frac{\overline{X}-\theta/2}{\sqrt{\theta^2/12n}}
$$

​的极限分布.

### 020319-解答

由于 $X_i\sim U(0,\theta)$, 则

$$
EX_i=\frac{\theta}{2},\quad DX_i=\frac{\theta^2}{12}.
$$

由于 CLT, 有

$$
\frac{\overline{X}-\theta/2}{\sqrt{\theta^2/12n}}\xrightarrow{\mathcal{L}}N(0,1).

## 020320

若 $X_1,X_2,\cdots,X_n$ 是来自 Poisson 分布总体 $P(\lambda)$, $\lambda>0$ 的样本, 求

$$
\frac{\overline{X}-\lambda}{\sqrt{\overline{X}/n}}
$$

​的极限分布.

### 020320-解答

由于 $X_i\sim P(\lambda)$, 则

$$
E(X_i)=\lambda,\quad D(X_i)=\lambda.
$$

由 CLT,

$$
\frac{\overline{X}-\lambda}{\sqrt{\lambda/n}}\xrightarrow{\mathcal{L}}N(0,1).
$$

依据 Khinchin 大数定律:

$$
\overline X\xrightarrow P\lambda,
$$

则

$$
\frac{\sqrt{\lambda/n}}{\sqrt{\overline{X}/n}}\xrightarrow{P}\frac{\sqrt{\lambda/n}}{\sqrt{\lambda/n}}=1,
$$

依据 Slutsky 定理,

$$
\frac{\overline{X}-\lambda}{\sqrt{\overline{X}/n}}=\frac{\overline{X}-\lambda}{\sqrt{\lambda/n}}\frac{\sqrt{\lambda/n}}{\sqrt{\overline{X}/n}}\xrightarrow{\mathcal{L}}\frac{\overline{X}-\lambda}{\sqrt{\lambda/n}}\sim N(0,1).
$$

## 020321

设 $X_1,X_2,\cdots,X_n$ 是来自均匀分布总体 $U(0,1)$ 的样本, $X_{(1)},\cdots,X_{(n)}$ 为顺序统计量,

(1)

若使 $P\{X_{(n)}\ge0.99\}$ 至少为 $0.95$, 则样本容量至少取多少?

(2)

求极差 $R_n=X_{(n)}-X_{(1)}$ 的密度函数;

(3)

当 $n$ 趋于无穷时, 求 $Z=2n(1-R_n)$ ​的极限分布.

### 020321-解答

(1)

由于 $X_1,X_2,\cdots,X_n\sim U(0,1)$,

则

$$
P(X_n\le x)=x.
$$

由于独立同分布的 $P(X_{(n)}\ge0.99)\ge0.95$, 独立同分布的性质 $X_1=X_2=s=X_n=X$, 则

$$
\begin{aligned}
P(X_{(n)}\ge0.99)&=1-P(X_{(n)}\le 0.99)\\
&=1-P(X\le 0.99)^n\\
&=1-0.99^n\ge0.95,
\end{aligned}
$$

则

$$
n\ge\left\lceil\frac{\ln0.05}{\ln0.99}\right\rceil=299.
$$

(2)

由分布函数的定义, $X_{(n)}$ 与 $X_{(1)}$ 的联合分布函数为

$$
\begin{aligned}
F(x_{(1)},x_{(n)})&=P\{X_{(1)}\le x_{(1)},X_{(n)}\le x_{(n)}\}\\
&=P\{X_{(n)}\le x_{(n)}\}-P\{X_{(1)}>x_{(1)}, X_{(n)}\le x_{(n)}\}\\
&=F(x_{(n)})^n-(F(x_{(n)})-F(x_{(1)}))^n,
\end{aligned}
$$

由于 $X_{(n)}$ 与 $X_{(1)}$ 未知是否独立, 故其联合密度函数为

$$
\begin{aligned}
&f(x_{(1)},x_{(n)})=\frac{\partial^2F(x_{(1)},x_{(n)})}{\partial x_{(1)}\partial x_{(n)}}\\
&=n(n-1)(F(x_{(n)})-F(x_{(1)}))^{n-2}
\end{aligned}
$$

由于 $\{X_i\}_{i=1}^n\sim U(0,1)$, 因此

$$
F(X\le x)=x.
$$

则联合密度函数为

$$
f(x_{(1)},x_{(n)})=n(n-1)(x_{(n)}-x_{(1)})^{n-2}.
$$

令

$$
\begin{cases}
R_n=X_{(n)}-X_{(1)},\\
S=X_{(1)},
\end{cases}
$$

则

$$
\begin{cases}
X_{(1)}=S,\\
X_{(n)}=S+R_n,
\end{cases}
$$

$$
|J|=\frac{\partial(x_{(1)},x_{(n)})}{\partial(s,r)}=1.
$$

因此

$$
f_{S,R_n}(s,r_n)=f_{X_{(1)},X_{(n)}}|J|=n(n-1)r_n^{n-2}.
$$

其中 $0<S<S+R_n<1$.

对 $S$ 积分, 求得 $R_n$ 的边缘密度:

$$
\begin{aligned}
f_{R_n}(r)&=\int_0^{1-r}n(n-1)r^{n-2}ds\\
&=n(n-1)r^{n-2}(1-r),
\end{aligned}
$$

其中 $0<r<1$.

(3)

由 (2) 知

$$
f_{R_n}(r)=n(n-1)r^{n-2}(1-r),
$$

其中 $0<r<1$. 令 $Z=2n(1-R_n)$, 则

$$
R_n=1-\frac{Z}{2n},\quad\left|\frac{dR_n}{dZ}\right|=\frac{1}{2n},
$$

因此

$$
\begin{aligned}
&\lim_{n\to\infty}f_Z(z)=\lim_{n\to\infty}f_{R_n}\left(1-\frac{z}{2n}\right)\frac{1}{2n}\\
&=\lim_{n\to\infty}\frac{(n-1)z}{4n}\left(1-\frac{z}{2n}\right)^{n-2}=\frac z4e^{-z/2}
\end{aligned}
$$

已知 $\chi^2$ 分布的密度函数

$$
f(x)=\frac{x^{n/2-1}e^{-x/2}}{2^{n/2}\Gamma(n/2)},
$$

因此 $Z\sim\chi^2(4)$.

## 0203X1

设总体 $X\sim N(\mu, \sigma^2)$, 样本 $X_1,\cdots,X_{2n}$, 且

$$
\overline{X}=\frac1{2n}\sum_{i=1}^{2n}X_i.
$$

证明

$$
\frac{y}{2\sigma^2}\sim\chi^2(n-1),\quad y=\sum_{i=1}^n(X_i+X_{n+i}-2\overline{X}^2).
$$

### 0203X1-解答

令 $Z_i=X_i+X_{n+i}$, 由于$X_1, X_2, \cdots, X_{2n} \sim N(0, 1)$ 独立, 则正态分布具有可加性,

$$
Z_i\sim N(2\mu, 2\sigma^2).
$$

$$
\overline Z=\frac1n\sum_{i=1}^nZ_i=\frac1n\sum_{i=1}^{2n}X_i=2\overline X,
$$

则

$$
y=\sum_{i=1}^n(X_i+X_{n+i}-2\overline{X})^2=\sum_{i=1}^n(Z_i-\overline Z)^2.
$$

对于 $X_i\sim N(\mu,\sigma^2)$, 由于 $\sum_{i=1}^n(X_i-\overline X)^2/\sigma^{2}\sim\chi^2(n-1)$, 则

$$
\frac{y}{2\sigma^2}\sim\chi^2(n-1).
$$

## 0203X2

设总体 $X\sim N(0, \sigma^2)$, 样本 $X_1$, $X_2$. 求

$$
F=\frac{(X_1+X_2)^2}{(X_1-X_2)^2}
$$

的分布.

### 0203X2-解答

由于 $X_1,X_2\sim N(0,\sigma^2)$ 独立, 因此正态分布具有可加性,

$$
X_1+X_2\sim N(0, 2\sigma^2),
$$

$$
X_1-X_2\sim N(0, 2\sigma^2).
$$

因此

$$
\left(\frac{X_1+X_2-0}{\sqrt2\sigma}\right)^2\sim\chi^2(1),
$$

$$
\left(\frac{X_1-X_2-0}{\sqrt2\sigma}\right)^2\sim\chi^2(1).
$$

因此

$$
\frac{(X_1+X_2)^2/1}{(X_1-X_2)^2/1}\sim F(1,1).
$$

验证 $X_1+X_2$ 与 $X_1-X_2$ 独立:

由于 $X_1,X_2$ 独立, 协方差的双线性性:

$$
\begin{aligned}
&\text{Cov}(X_1+X_2,X_1-X_2)\\
&=\text{Cov}(X_1,X_1)-\text{Cov}(X_1,X_2)\\
&+\text{Cov}(X_2,X_1)-\text{Cov}(X_2,X_2)\\
&=D(X_1)-D(X_2)=0.
\end{aligned}
$$
