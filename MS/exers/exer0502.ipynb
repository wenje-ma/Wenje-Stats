## 050205

从标准差为 $\sigma=5.2$ 的正态总体 $N(\mu,\sigma^2)$ 中随机抽取容量为 16 的样本, 计算得样本均值为 $\overline{x}=27.56$. 在显著性水平 $\alpha=0.05$ 下, 能否认为总体均值 $\mu=26$?

### 050205-解答

提出假设

$$
H_0:\mu=26,\quad H_1:\mu\ne26.
$$

总体方差已知, 因此构造检验统计量

$$
Z=\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}\sim N(0,1).
$$

查表得 $u_{0.025}=1.96$, 因此拒绝域为

$$
D=\left\{\left|\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}\right|\ge1.96\right\}.
$$

计算得

$$
\frac{|\overline{x}-\mu_0|}{\sigma/\sqrt{n}}=1.2<1.96.
$$

无法拒绝 $H_0$, 即认为 $\mu=26$ 是合理的.

## 050207

要求一种元件的平均使用寿命不得低于 $1000\,\mathrm{h}$, 生产者从这种元件中随机抽取 25 件, 测得其寿命的平均值为 950h. 已知该种元件寿命服从标准差为 $\sigma=100\,\mathrm{h}$ 的正态分布, 在显著性水平 $\alpha=0.05$ 下, 判断这批元件是否合格?

### 050207-解答

提出假设

$$
H_0:\mu\ge1000;\quad H_1:\mu<1000.
$$

总体方差已知, 因此构造检验统计量

$$
Z=\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}\sim N(0,1).
$$

查表得 $u_{0.05}=1.645$, 因此拒绝域为

$$
D=\left\{\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}\le-1.645\right\}.
$$

计算得

$$
\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}=-2.5\le-1.645,
$$

拒绝 $H_0$, 即认为这批元件不合格.

## 050209

设 $X_1,X_2,\cdots,X_{25}$ 是来自正态总体 $N(\mu,3^2)$ 的一个样本, 其中 $\mu\in\mathbb{R}$ 为未知参数. 设检验问题

$$
H_0:\mu=\mu_0;\quad H_1:\mu\ne\mu_0
$$

($\mu_0$ 已知)的拒绝域为

$$
|\overline{X}-\mu_0|\ge c.
$$

(1)

在显著性水平 $\alpha=0.05$ 下, 求常数 $c$;

(2)

求 $\mu=\mu_1(\mu_1\ne\mu_0)$ 时犯第二类错误的概率(用标准正态分布函数 $\varPhi()$ 表示).

### 050209-解答

(1)

总体方差已知, 因此构造检验统计量

$$
Z=\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}\sim N(0,1),
$$

查表得 $u_{0.025}=1.96$, 因此拒绝域为

$$
D=\left\{\left|\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}\right|\ge1.96\right\},
$$

即 $|\overline{X}-\mu_0|\ge1.176$, $c=1.176$.

(2)

由(1)知, 

$$
\begin{aligned}
\beta&=P\{1-P(|Z|\ge1.96)|\mu=\mu_1\}\\
&=P\{|Z|\le1.96|\mu=\mu_1\}.
\end{aligned}
$$

由于 $X\sim N(\mu_1,3^2)$, 则

$$
\begin{aligned}
\frac{\overline{X}-\mu_1}{\sigma/\sqrt{n}}&=\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}+\frac{\mu_0-\mu_1}{\sigma/\sqrt{n}}\\
&=Z+\frac53(\mu_0-\mu_1)\sim N(0,1),
\end{aligned}
$$

因此

$$
P(|Z|<1.96)=\Phi\left(1.96+\frac53(\mu_0-\mu_1)\right)-\Phi\left(-1.96+\frac53(\mu_0-\mu_1)\right).
$$

## 050212

已知维尼纶纤度在正常情况下服从正态分布 $N(\mu,\sigma^2)$, 且 $\sigma=0.048$. 从某天的产品中任取 5 根纤维, 测得其纤度为

$$
1.32,\,1.55,\,1.36,\,1.40,\,1.44,
$$

在显著性水平 $\alpha=0.05$ 下, 问这天的标准差是否正常?

### 050212-解答

提出假设:

$$
H_0:\sigma=0.048;\quad H_1:\sigma\ne0.048.
$$

$\overline{x}=1.414$, $s^2=7.78\times10^{-3}$, 令

$$
Z=\frac{(5-1)S^2}{\sigma^2}\sim\chi^2(4),
$$

查表得 $\chi^2_{0.025}(4)=11.143,\chi^2_{0.975}(4)=0.484$, 因此拒绝域为

$$
D=\left\{\frac{4S^2}{\sigma^2}\ge11.143\text{ or }\frac{4S^2}{\sigma^2}\le0.484\right\}.
$$

计算得

$$
\frac{4s^2}{\sigma^2}=13.5070\ge11.143,
$$

拒绝 $H_0$, 即标准差不正常.

## 050215

测量某物体的重量 10 次, 得样本均值为 $\overline{x}=10.05$, 样本方差为 $s^2=0.0576$. 假定该物体的重量服从正态分布 $N(\mu,\sigma^2)$.

(1)

若 $\sigma^2=0.2^2$, 在显著性水平 $\alpha=0.05$ 下, 检验

$$
H_0:\mu=10;\quad H_1:\mu>10.
$$

(2)

若 $\sigma^2$ 未知, 在显著性水平 $\alpha=0.05$ 下, 检验

$$
H_0:\mu=10;\quad H_1:\mu>10.
$$

(3)

在显著性水平 $\alpha=0.05$ 下, 检验

$$
H_0:\sigma^2\ge0.25^2;\quad H_1:\sigma^2<0.25^2.
$$

### 050215-解答

(1)

总体方差已知, 因此构造检验统计量

$$
Z=\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}\sim N(0,1).
$$

查表得 $u_{0.05}=1.645$, 因此拒绝域为

$$
D=\left\{\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}>1.645\right\}.
$$

计算得

$$
\frac{\overline{x}-\mu_0}{\sigma/\sqrt{n}}=0.7906<1.645.
$$

无法拒绝 $H_0$, 即不能认为 $\mu>10$.

(2)

总体方差未知, 因此构造检验统计量

$$
Z=\frac{\overline{X}-\mu_0}{S/\sqrt{n}}\sim t(9).
$$

查表得 $t_{0.05}(9)=1.8331$, 因此拒绝域为

$$
D=\left\{\frac{\overline{X}-\mu_0}{S/\sqrt{n}}>1.8331\right\}.
$$

计算得

$$
\frac{\overline{x}-\mu_0}{s/\sqrt{n}}=0.6588<1.8331.
$$

无法拒绝 $H_0$, 即不能认为 $\mu>10$.

(3)

令

$$
Z=\frac{9S^2}{\sigma_0^2}\sim\chi^2(9).
$$

查表得 $\chi^2_{0.95}(9)=3.325$, 因此拒绝域为

$$
D=\left\{\frac{9S^2}{\sigma_0^2}<3.325\right\}.
$$

计算

$$
\frac{9s^2}{\sigma_0^2}=8.2944>3.325,
$$

无法拒绝 $H_0$, 即认为 $\sigma^2\ge0.25^2$.

## 050217

比较甲、乙两种棉花品种的优劣, 假设用它们纺出的棉纱强度分别服从正态分布

$$
N(\mu_1,\sigma_1^2),\quad N(\mu_2,\sigma_2^2),
$$

试验者分别从这两种棉纱中抽取样本 $X_1,X_2,\cdots,X_{200}$, $Y_1,Y_2,\cdots,Y_{100}$, 得样本均值和样本方差分别为

$$
\overline{x}=5.6,\quad\overline{y}=5.2,\quad s_1^2=4,\quad s_2^2=2.56.
$$

(1)

若 $\sigma_1^2=2.2^2,\sigma_2^2=1.8^2$, 在显著性水平 $\alpha=0.05$ 下, 检验

$$
H_0:\mu_1\le\mu_2;\quad H_1:\mu_1>\mu_2.
$$

(2)

若 $\sigma_1^2=\sigma_2^2$ 未知, 在显著性水平 $\alpha=0.05$ 下, 检验

$$
H_0:\mu_1\le\mu_2;\quad H_1:\mu_1>\mu_2.
$$

### 050217-解答

(1)

构造检验统计量

$$
Z=\frac{\overline{X}-\overline{Y}-(\mu_1-\mu_2)}{\sqrt{\sigma_1^2/n_1+\sigma_2^2/n_2}}\sim N(0,1).
$$

拒绝域 $D=\{Z>u_{0.05}\}$, 代入数据计算得

$$
Z\approx1.189<1.645=u_{0.05},
$$

无法拒绝 $H_0$, 则认为 $\mu_1\le\mu_2$.

(2)

构造检验统计量:

$$
T=\frac{\overline{X}-\overline{Y}-(\mu_1-\mu_2)}{S_w\sqrt{1/n_1+1/n_2}}\sim t(n_1+n_2-2),
$$

$$
S_\omega^2=\frac{(n_1-1)S_1^2+(n_2-1)S_2^2}{n_1+n_2-2}.
$$

拒绝域 $D=\{T>t_{0.05}\}$, 代入数据计算得

$$
T\approx1.231<1.645,
$$

无法拒绝 $H_0$, 则认为 $\mu_1\le\mu_2$.

## 050221

对两批同类电子元件的电阻(单位: $\Omega$)进行测试, 各抽取6件测得结果如下:

第一批:

$$
0.140,\,0.138,\,0.143,\,0.142,\,0.144,\,0.137;
$$

第二批:

$$
0.135,\,0.140,\,0.142,\,0.136,\,0.138,\,0.140.
$$

已知两批元件的电阻分别服从正态分布 $N(\mu_1,\sigma_1^2)$ 和 $N(\mu_2,\sigma_2^2)$, 在显著性水平 $\alpha=0.05$ 下, 分别检验:

(1)

两批元件电阻的方差是否相等;

(2)

两批元件的平均电阻有无显著差异.

### 050221-解答

(1)

$$
\begin{align*}
\overline{x}=0.1407,\quad&s_1^2=7.866\times10^{-5}\\
\overline{y}=0.1385,\quad&s_2^2=7.140\times10^{-5},
\end{align*}
$$

提出假设:

$$
H_0:\sigma_1^2=\sigma_2^2;\quad H_1:\sigma_1^2\ne\sigma_2^2.
$$

构造检验统计量:

$$
F=\frac{S_1^2/S_2^2}{\sigma_1^2/\sigma_2^2}\sim F(5,5),
$$

拒绝域

$$
D=\{F\ge F_{0.025}(5,5)\text{ or }F\le F_{0.975}(5,5)\},
$$

计算

$$
F=1.108\in[0.14,7.15],
$$

无法拒绝 $H_0$, 即认为方差相等.

(2)

提出假设

$$
H_0:\mu_1=\mu_2,H_1:\mu_1\ne\mu_2.
$$

构造检验统计量

$$
T=\frac{\overline{X}-\overline{Y}}{S_w\sqrt{1/n_1+1/n_2}},
$$

$$
S_\omega^2=\frac{(n_1-1)S_1^2+(n_2-1)S_2^2}{n_1+n_2-2}=7.484\times10^{-5},
$$

拒绝域 $D=\{|T|>t_{0.025}(18)\}$, 计算得

$$
|T|=1.372<2.228,
$$

无法拒绝 $H_0$, 即认为平均电阻无明显差异.
