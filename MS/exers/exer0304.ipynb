## 030401

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本, 求 $c$ 使得

$$
c\sum_{i=1}^{n-1}(X_{i+1}-X_i)^2
$$

是 $\sigma^2$ 的无偏估计.

### 030401-解答

由于独立同分布 $X_i \sim N(\mu,\sigma^2)$, 则正态分布具有可加性,

$$
X_{i+1}-X_i \sim N(0,2\sigma^2),
$$

因此

$$
\begin{aligned}
&E(X_{i+1}-X_i)^2=D(X_{i+1}-X_i)\\
&+(E(X_{i+1}-X_i))^2=2\sigma^2,
\end{aligned}
$$

$$
E\sum_{i=1}^{n-1} (X_{i+1}-X_i)^2=2(n-1)\sigma^2,
$$

则 $c=1/(2(n-1))$.

## 030405

设 $X_1,X_2,\cdots,X_n$ 是来自总体 $X$ 的样本, 且 $X$ 的密度函数为

$$
f(x,\theta)=
\begin{cases}
e^{-(x-\theta)},&x>\theta,
\\0,&\text{Others},
\end{cases}$$

(1)

求参数 $\theta$ 的矩估计 $\widehat{\theta}_1$, 并判断 $\widehat{\theta}_1$ 是否是 $\theta$ 的无偏估计;

(2)

求参数 $\theta$ 的极大似然估计 $\widehat{\theta}_2$, 并判断 $\widehat{\theta}_2$ 是否是 $\theta$ 的无偏估计?

### 030405-解答

(1)

令 $Y=X-\theta$, 则 $Y \sim E(1)$, 因此 $EY=1$, $DY=1$.

$$
\overline{X}=EX=E(Y+\theta)=1+\theta,
$$

则 $\widehat{\theta_1}=\overline{X}-1$.

$$
E(\widehat{\theta_1})=E\overline{X}-1=\theta,
$$

是无偏估计.

(2)

$$
L(\theta)=
\begin{cases}
\exp\{-\sum x_i+n\theta\},&X_{(1)}>\theta,\\
0,&\text{Others},
\end{cases}
$$

$L(\theta)$ 单增, 故 $\widehat{\theta_2}=X_{(1)}$. 已知

$$
F(x)=
\begin{cases}
1-e^{-(x-\theta)},&x>\theta,\\
0,&x\le\theta,
\end{cases}
$$

则

$$
F_{X_{(1)}}(x)=1-(1-F(x))^n.
$$

因此

$$
f_{X_{(1)}}(x)=n(1-F(x))^{n-1}f(x)=ne^{-n(x-\theta)},
$$

即

$$
X_{(1)}-\theta \sim \exp\{1/n\},
$$

$$
E(X_{(1)})=\theta+n\ne\theta,
$$

不是无偏估计.

## 030407

设 $X_1,X_2$ 是来自总体 $X$ 的样本, 且 $X$ 的密度函数为

$$
f(x,\theta)=
\begin{cases}
k\theta^k x^{-(k+1)},&x>\theta,\\
0,&\text{Others},
\end{cases}
$$

其中 $k>2$, $\theta>0$ 为未知参数.

(1)

证明

$$
T_1=\frac{k-1}{2k}(X_1+X_2)
$$

和

$$
T_2=\frac{2k-1}{2k}X_{(1)}
$$

都是 $\theta$ 的无偏估计;

(2)

指出 $T_1$ 和 $T_2$ 哪一个有效.

### 030407-解答

(1)

$$
\begin{aligned}
EX=\int_{\theta}^{\infty}xk\theta^kx^{-(k+1)}dx=\frac{k\theta}{k-1},
\end{aligned}
$$

$$
E(T_1)=\frac{k-1}{2k}  2  \frac{k\theta}{k-1}=\theta,
$$

是无偏估计.

$$
F_{X_{(1)}}(x)=1-(1-F(x))^n,
$$

$$
F(x)=\int_{\theta}^{x}k\theta^kt^{-(k+1)}dt=1-\left(\frac{\theta}{x}\right)^k,
$$

其中 $x>0$. 因此

$$
f_{X_{(1)}}(x)=n(1-F(x))^{n-1}f(x)=2k\frac{\theta^{2k}}{x^{2k+1}},\quad x>\theta.
$$

$$
\begin{aligned}
E(X_{(1)})&=\int_{\theta}^{\infty} x  2k \frac{\theta^{2k}}{x^{2k+1}} dx\\
&=2k\theta^{2k}\int_{\theta}^{\infty} x^{-2k}dx\\
&=\frac{2k\theta}{2k-1}.
\end{aligned}
$$

$$
E(T_2)=\frac{2k-1}{2k}\frac{2k\theta}{2k-1}=\theta,
$$

是无偏估计.

(2)

$$
EX^2=\int_{0}^{\infty}x^2k\theta^kx^{-(k+1)}dx=\frac{k\theta^2}{k-2},
$$

$$
\begin{aligned}
DX&=EX^2-(EX)^2\\
&=k\theta^2\left(\frac{1}{k-2}-\frac{k}{(k-1)^2}\right)\\
&=\frac{k\theta^2}{(k-1)^2(k-2)},
\end{aligned}
$$

$$
\begin{aligned}
D(T_1)&=\left(\frac{k-1}{2k}\right)^2(DX_1+DX_2)\\
&=\frac{(k-1)^2}{2k^2}D(X)\\
&=\frac{\theta^2}{2k(k-2)},
\end{aligned}
$$

$$
\begin{aligned}
EX_{(1)}^2&=\int_{\theta}^{\infty} x^22k\theta^{2k}x^{-(2k+1)}dx\\
&=2k\theta^{2k}\int_{\theta}^{\infty} x^{-2k+1}dx\\
&=\frac{k\theta^2}{k-1},
\end{aligned}
$$

$$
\begin{aligned}
D(X_{(1)})&=EX_{(1)}^2-(EX_{(1)})^2\\
&=\frac{k\theta^2}{k-1}-\frac{4k^2\theta^2}{(2k-1)^2}\\
&=\frac{k\theta^2}{(k-1)(2k-1)^2}
\end{aligned}
$$

$$
\begin{aligned}
D(T_2)&=\left(\frac{2k-1}{2k}\right)^2 D(X_{(1)})\\
&=\frac{\theta^2}{4k(k-1)}.
\end{aligned}
$$

$k>2$, 则 $D(T_1)>D(T_2)$, $T_2$ 更有效.

## 030410

设 $\widehat{\theta}_n$ 是 $\theta$ 的估计量, 证明: 若 $n\to+\infty$ 时,

$$
E(\widehat{\theta}_n)\to\theta
$$

$$
D(\widehat{\theta}_n)\to0,
$$

则 $\widehat{\theta}_n$ 是 $\theta$ 的相合估计.

### 030410-解答

只需证

$$
\lim_{n\to\infty} P(|\widehat{\theta}_n-\theta|\ge\varepsilon)=0.
$$

由绝对值不等式:

$$
|\widehat{\theta}_n-\theta|\le |\widehat{\theta}_n-E(\widehat{\theta}_n)|+|E(\widehat{\theta}_n)-\theta|,
$$

由于 $E(\widehat{\theta}_n)\to\theta$, 则存在 $\varepsilon>0$ 使得

$$
|E(\widehat{\theta}_n)-\theta|<\frac{\varepsilon}{2}.
$$

由 $D(\widehat{\theta}_n)\to0$ 与 Chebyshev 不等式:

$$
P\left\{|\widehat{\theta}_n-E(\widehat{\theta}_n)|\ge\frac{\varepsilon}{2}\right\}\le\frac{D(\widehat{\theta}_n)}{\varepsilon^2/2}\to 0,
$$

## 030412

设 $X_1,X_2,\cdots,X_n$ 是来自均匀分布 $U[0,\theta]$, $\theta>0$ 的样本, 证明

$$
\sqrt[n]{X_1X_2\cdots X_n}
$$

为 $\theta/e$ 的相合估计.

### 030412-解答

令 $Y_i=\ln X_i$, 即证

$$
\frac{1}{n}\sum Y_i\xrightarrow{P} \ln\theta-1,
$$

由 Khinchin 大数定律:

$$
\frac{1}{n}\sum Y_i \xrightarrow{P} E(Y_i),
$$

只需证 $E(Y_i)=\ln\theta -1$.

$$
\begin{aligned}
E(Y_i)&=E(\ln X_i)\\
&=\int_{0}^{\theta}\ln x\frac{1}{\theta}dx\\
&=\ln\theta-1.
\end{aligned}
$$
