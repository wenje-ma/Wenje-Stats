## 040301

设某电子产品的寿命服从指数分布 $E(\lambda)$, 其密度函数为

$$
f(x)=\begin{cases}
\lambda\text{e}^{-\lambda x},&x>0,\\
0,&\text{Others},
\end{cases}$$

其中 $\lambda>0$. 现从该分布的一批产品中抽取容量为 $9$ 的样本, 测得寿命为(单位: kh):

$$
15,\,45,\,50,\,53,\,60,\,65,\,70,\,83,\,90.
$$

求平均寿命 $1/\lambda$ 的置信水平为 90% 的置信区间和置信上限, 下限.

### 040301-解答

$$
\frac{2n\overline{X}}{\lambda}\sim\chi^2(2n),
$$

查表得 $\chi^2_{0.05}(18)=28.869$, $\chi^2_{0.95}(18)=9.390$, 因此

$$
P\left\{9.390<\frac{2\times1859}{\lambda}<28.869\right\}=0.9,
$$

置信区间为 $[36.787,113.099]$.

查表得 $\chi^2_{0.1}(18)=25.989$, $\chi^2_{0.9}(18)=10.865$,

因此置信上限为 $97.745$,下限为 $40.863$.

## 040303

设 $X_1,X_2,\cdots,X_n$ 是来自总体 $X$ 的样本, 且 $X$ 的密度函数为

$$
f(x,\theta)=
\begin{cases}
\theta/x^{2},&x>\theta,\\
0,&\text{Others},
\end{cases}
$$

其中 $\theta>0$. 求参数 $\theta$ 的置信水平为 $1-\alpha$ 的置信区间.

### 040303-解答

注意到定义域中出现了参数, 因此可类比均匀分布考虑顺序统计量: 依据因子分解定理,

$$
\prod_{i=1}^nf(x,\theta)=\frac{\theta^n}{\prod_{i=1}^nx^2}I_{\{X_{(1)}>\theta\}},
$$

即 $X_{(1)}$ 是充分统计量. $X_{(1)}$ 的分布函数

$$
\begin{aligned}
&F_{X_{(1)}}(x)=1-(1-F(x))^n\\
&=1-(1-(1-\theta/x))^n\\
&=1-\theta^n/x^n,
\end{aligned}
$$

$X_{(1)}$ 的密度函数为

$$
f_{X_{(1)}}(x)=\frac{n\theta^n}{x^{n+1}}.
$$

需要构造枢轴量使得其分布与 $\theta$ 无关, 设枢轴量为 $T$, 则

$$
\begin{aligned}
f_T(t)&=f_{X_{(1)}}(x)\frac{dx}{dt}\\
&=\frac{n\theta^n}{x^{n+1}}\frac{dx}{dt}\\
&=\frac{n}{\theta(x/\theta)^{n+1}}\frac{dx}{dt}.
\end{aligned}
$$

令 $t=x/\theta$, 则 $dx/dt=\theta$, 发现

$$
f_T(t)=\frac n{t^{n+1}}
$$

与 $\theta$ 无关, 因此枢轴量 $T=X_{(1)}/\theta$, 其置信水平满足

$$
P\{t_1\le X_{(1)}/\theta\le t_2\}=1-\alpha.
$$

由密度函数 $f_T(t)$ 积分求得其分布函数 $F_T(t)$ 为

$$
F_T(t)=
\begin{cases}
1-1/t^n,&t\ge1,\\
0,&t<1,
\end{cases}
$$

因此

$$
\frac1{t_1^n}-\frac1{t_2^n}=1-\alpha.
$$

要求 $t_2-t_1$ 的最小值, 使用 Lagrange 乘数法: 设

$$
\mathcal L=(t_2-t_1)+\lambda\left(\frac1{t_1^n}-\frac1{t_2^n}-(1-\alpha)\right)
$$

令 $\mathcal L$ 分别对 $t_1$, $t_2$ 求偏导为 $0$, 解得 $t_1=t_2$ 矛盾, 这说明区间 $[t_1,t_2]$ 在边界中取到.

由于 $t\ge1$, 因此 $t_1=1$ ($t_2=\infty$ 舍去), $t_2=\alpha^{-n}$, 即

$$
1\le X_{(1)}/\theta\le\alpha^{-1/n},
$$

$$
\alpha^{1/n}X_{(1)}\le\theta\le X_{(1)},
$$

置信区间为 $[\alpha^{1/n}X_{(1)},X_{(1)}]$.

## 040304

设 $X_1, X_2, \cdots, X_n$ 是来自总体 $X$ 的样本, 且 $X$ 的密度函数为

$$
f(x)=
\begin{cases}
2\lambda x\text{e}^{-\lambda x^2},&x>0,\\
0,&\text{Others},
\end{cases}$$

其中 $\lambda>0$. 求参数 $\lambda$ 的置信水平为 $1-\alpha$ 的置信区间.

### 040304-解答

观察密度函数:

$$
\begin{aligned}
f(x)&=2\lambda xe^{-\lambda x^2}I_{\{x>0\}}\\
&=\frac{e^{-\lambda x^2}d(\lambda x^2)}{dx}I_{\{x>0\}},
\end{aligned}
$$

因此令 $Y=\lambda X^2$.

$$
\begin{aligned}
F_Y(y)&=P(\lambda X^2\le y)\\
&=P\left(0<X\le\sqrt{\frac{y}{\lambda}}\right)\\
&=\int_{0}^{\sqrt{y/\lambda}}\frac{e^{-\lambda x^2}d(\lambda x^2)}{dx}dx \\
&=1-e^{-y},\quad y>0,
\end{aligned}
$$

因此 $Y\sim E(1)$,

$$
2\lambda\sum X_i^2\sim\chi^2(2n).
$$

$$
P\{\chi^2_{1-\alpha/2}(2n)\le2\lambda\sum X_i^2\le\chi^2_{\alpha/2}(2n)\}=1-\alpha,
$$

$\lambda$ 的置信区间为

$$
\left[\frac{\chi^2_{1-\alpha/2}(2n)}{2\sum X_i^2},\frac{\chi^2_{\alpha/2}(2n)}{2\sum X_i^2}\right].
$$

## 040310

在一批货物中随机抽取 $100$ 件, 发现有 $16$ 件不合格品, 求这批货物的不合格品率的置信水平近似为 $95\%$ 的置信区间.

### 040310-解答

不合格品总数 $X\sim B(n,p)$,

$$
E(X)=np,\quad DX=np(1-p).
$$

则不合格品率 $\widehat{p}=X/n$,

$$
E(\widehat{p})=p,\quad D(\widehat{p})=\frac{p(1-p)}{n}.
$$

依据CLT, 大样本下

$$
\frac{\widehat{p}-p}{\sqrt{p(1-p)/n}}\sim N(0,1),
$$

$$
P\left\{-z_{\alpha/2}\le\frac{\widehat{p}-p}{\sqrt{p(1-p)/n}}\le z_{\alpha/2}\right\}=1-\alpha,
$$

解得 $p$ 的置信区间为

$$
\left[\widehat{p}-z_{\alpha/2}\sqrt{\frac{\widehat{p}(1-\widehat{p})}{n}},\widehat{p}+z_{\alpha/2}\sqrt{\frac{\widehat{p}(1-\widehat{p})}{n}}\right].
$$

代入数据得区间为 $[0.088,0.232]$.
