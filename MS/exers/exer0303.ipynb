## 030303

设 $X_1,X_2,\cdots,X_n$ 是来自总体 $X$ 的样本, 且 $X$ 的密度函数 $f(x)$ 如下所示, 求其中未知参数的极大似然估计.

(1)

$$
f(x,\theta)=
\begin{cases}
(\theta+1)x^\theta,&0<x<1,\\
0,&\text{Others},
\end{cases}
$$

其中 $\theta>-1$ 为未知参数;

(2)

$$
f(x,\theta)=
\begin{cases}
\theta\exp\{-\theta x\},&x>0,\\
0,&x\le0,
\end{cases}
$$

其中 $\theta>0$ 为未知参数;

(3)

$$
f(x,\theta)=
\frac{1}{2\theta}e^{-\frac{|x|}{\theta}},
$$

其中 $x\in\mathbb{R}$, $\theta>0$ 为未知参数;

(4)

$$
f(x,\theta)=
\begin{cases}1,&\theta-0.5\le x\le\theta+0.5, \\
0,&\text{Others},
\end{cases}
$$

其中 $\theta\in\mathbb{R}$ 为未知参数;

(5)

$$
f(x,\mu,\sigma^2)=
\begin{cases}
\frac{1}{\sqrt{2\pi}\sigma x}e^{-\frac{(\ln x-\mu)^2}{2\sigma^2}},&x>0,\\
0,&x\le0,
\end{cases}
$$

其中 $\mu\in\mathbb{R},\sigma^2>0$ ​为未知参数.

### 030303-解答

(1)

$$
L(\theta)=(\theta+1)^n\left(\prod_{i=1}^nX_i\right)^{\theta},
$$

$$
\ln L(\theta)=n\ln(\theta+1)+\theta\sum_{i=1}^n\ln X_i,
$$

$$
\ln L(\theta)'=\frac{n}{\theta+1}+\sum_{i=1}^n\ln X_i=0,
$$

$$
\widehat{\theta}=-\frac{n}{\sum_{i=1}^n\ln X_i}-1.
$$

(2)

$$
L(\theta)=\theta^n\exp\left\{-\theta\sum_{i=1}^nX_i\right\},
$$

$$
\ln L(\theta)=n\ln\theta-\theta\sum_{i=1}^nX_i,
$$

$$
\ln L(\theta)'=\frac{n}{\theta}-\sum_{i=1}^nX_i=0,
$$

$$
\widehat{\theta}=\frac{n}{\sum_{i=1}^nX_i}=\frac{1}{\overline{X}}.
$$

(3)

$$
L(\theta)=\frac{1}{(2\theta)^n}e^{-\frac{1}{\theta}\sum_{i=1}^n|X_i|},
$$

$$
\ln L(\theta)=-n\ln(2\theta)-\frac{1}{\theta}\sum_{i=1}^n|X_i|,
$$

$$
\ln L(\theta)'=-\frac{n}{\theta}+\frac{1}{\theta^2}\sum_{i=1}^n|X_i|=0,
$$

$$
\widehat{\theta}=\frac{1}{n}\sum_{i=1}^n|X_i|.
$$

(4)

直接观察密度函数图像可知, 当

$$
L(\theta)=\prod_{i=1}^nf(x_i,\theta)
$$

取最大值时, $\theta\in[X_{(n)}-0.5, X_{(1)}+0.5]$ 任一值.

(5)

$$
L(\mu,\sigma^2)=\prod_{i=1}^n\frac{1}{\sqrt{2\pi}\sigma x_i}\exp\left\{-\frac{(\ln X_i-\mu)^2}{2\sigma^2}\right\},
$$

$$
\begin{aligned}
\ln L(\mu,\sigma^2)&=-n\ln\sqrt{2\pi}-n\ln\sigma\\
&-\sum_{i=1}^n\ln X_i-\frac{1}{2\sigma^2}\sum_{i=1}^n(\ln X_i-\mu)^2,
\end{aligned}
$$

令 $Y_i=\ln X_i$, $C=-n\ln\sqrt{2\pi}-\sum_{i=1}^n\ln X_i$,

则

$$
\ln L(\mu,\sigma^2)=C-n\ln\sigma-\frac{1}{2\sigma^2}\sum_{i=1}^n(Y_i-\mu)^2.
$$

$$
\begin{cases}
\ln L(\mu,\sigma^2)'_{\mu}=\sum_{i=1}^n(Y_i-\mu)/\sigma^2=0,\\
\ln L(\mu,\sigma^2)'_{\sigma^2}=-n/(2\sigma^2)+\sum_{i=1}^n(Y_i-\mu)^2/(2\sigma^4)=0,
\end{cases}$$

$$
\begin{cases}
\widehat{\mu}=\overline{\ln X},\\
\widehat{\sigma}^2=\sum_{i=1}^n(\ln X_i-\overline{\ln X})^2/n.
\end{cases}
$$

## 030304

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本, 其中 $\mu\in\mathbb{R},\sigma>0$,

(1)

若 $\mu$ 未知, $\sigma$ 已知, 求 $\mu$ 的极大似然估计;

(2)

若 $\mu$ 已知, $\sigma$ 未知, 求 $\sigma^2$ ​的极大似然估计.

### 030304-解答

由 030303 (5) 知, $X\sim N(\mu,\sigma^2)$ 的极大似然估计:

$$
\begin{cases}
\widehat{\mu}=\overline{X},\\
\widehat{\sigma}^2=\sum_{i=1}^n(X_i-\overline{X})^2/n.
\end{cases}
$$

(1)

$\sigma^2$ 已知, 则 $\widehat{\mu}=\overline{X}$.

(2)

$\mu$ 已知, 则 $\widehat{\sigma}^2=\sum_{i=1}^n(X_i-\mu)^2/n$.

## 030306

设 $X_1,X_2,\cdots,X_n$ 是来自双参数指数分布总体的样本, 且密度函数为

$$
f(x,\mu,\lambda)=
\begin{cases}
\lambda\exp\{-\lambda(x-\mu)\},&x>\mu,\\
0,&\text{Others},
\end{cases}
$$

其中 $\mu\in\mathbb{R}$, $\lambda>0$ 为未知参数, 分别求 $\mu$, $\lambda$ 和

$$
\alpha=P\{X_1\ge t\}
$$​

的极大似然估计, 其中 $ t>\mu$.

## 030306-解答

$$
L(\mu,\lambda)=\lambda^ne^{-\lambda\sum_{i=1}^n(x_i-\mu)},\mu<X_{(1)},
$$

$$
\ln L(\mu,\lambda)=n\ln\lambda-\lambda\sum_{i=1}^nX_i+n\lambda\mu,
$$

$$
\ln L(\mu,\lambda)'_\mu=n\lambda>0,\quad\forall\mu,
$$

$$
\ln L(\mu,\lambda)'_\lambda=\frac{n}{\lambda}-(\sum_{i=1}^nX_i-n\widehat{\mu})=0
$$

解得

$$
\begin{cases}
\widehat{\mu}=X_{(1)},\\
\widehat{\lambda}=n/\sum_{i=1}^n(x_i-X_{(1)}),
\end{cases}
$$

$$
P\{X\ge t\}=\int_{t}^{\infty}\lambda e^{-\lambda(x-\mu)}dx=e^{-\lambda(t-\mu)},
$$

$$
\widehat{\alpha}=\exp\left\{-\frac{n(t-X_{(1)})}{\sum(X_i-X_{(1)})}\right\}.
$$
