## 030203

设 $X_1,X_2,\cdots,X_n$ 是来自总体 $X$ 的样本, 且 $X$ 的密度函数 $f(x)$ 如下所示, 求其中未知参数的矩估计.

(1)

$$
f(x,\theta)=
\begin{cases}
(\theta+1)x^\theta,&0<x<1,\\
0,&\text{Others},
\end{cases}
$$

其中 $\theta>-1$ 为未知参数;

(2)

$$
f(x,\theta)=
\begin{cases}
\theta\exp\{-\theta x\},&x>0,\\
0,&x\le0,
\end{cases}
$$

其中 $\theta>0$ 为未知参数;

(3)

$$
f(x,\theta)=\frac{1}{2\theta}e^{-\frac{|x|}{\theta}},
$$,

其中 $x\in\mathbb{R}$, $\theta>0$ 为未知参数$;

(4)

$$
f(x,\theta)=
\begin{cases}
1,&\theta-0.5\le x\le\theta+0.5,\\
0,&\text{Others},
\end{cases}
$$

其中 $\theta\in\mathbb{R}$ 为未知参数;

(5)

$$
f(x,\mu,\sigma^2)=
\begin{cases}
\exp\{-((\ln x-\mu)^2)/(2\sigma^2)\}/(\sqrt{2\pi}\sigma x),&x>0,\\
0,&x\le0,\end{cases}
$$

其中 $\mu\in\mathbb{R}$, $\sigma>0$ 为未知参数;

(6)

$$
f(x,\mu,\lambda)=\begin{cases}
\lambda\exp\{-\lambda(x-\mu)\}, & x>\mu, \\
0, & \text{Others},
\end{cases}$$

其中 $\mu\in\mathbb{R}$, $\lambda>0$ 为未知参数.

### 030203-解答

(1)

$$
EX=\int_{0}^{1}x(\theta+1)x^{\theta}dx=\frac{\theta+1}{\theta+2},
$$

令 $EX=\overline{X}$, 则

$$
\widehat{\theta}=\frac{2\overline{X}-1}{1-\overline{X}}.
$$

(2)

由于 $X\sim E(\theta)$, 则 $EX=1/\theta$, 令 $EX=\overline{X}$, 则

$$
\widehat{\theta}=\frac{1}{\overline{X}}.
$$

(3)

$$
EX=\int_{-\infty}^{+\infty}x\frac{1}{2\theta}e^{-\frac{|x|}{\theta}}dx=0,
$$

发现其不含 $\theta$, 故用 $E|X|$:

$$
\begin{aligned}
E|X|&=\int_{-\infty}^{+\infty}|x|\frac{1}{2\theta}e^{-\frac{|x|}{\theta}}dx\\
&=2\int_0^{+\infty}x\frac{1}{2\theta}e^{-\frac{x}{\theta}}dx\\
&=\theta\int_0^{+\infty}ue^{-u}du=\theta.
\end{aligned}
$$

令 $E|X|=|\overline{X}|$, 则 $\widehat{\theta}=|\overline{X}|$.

(4)

$$
EX=\int_{\theta-0.5}^{\theta+0.5}xdx=\theta,
$$

令 $EX=\overline{X}$, 则 $\widehat{\theta}=\overline{X}$.

(5)

令 $Y=\ln X\in\mathbb R$, 则 $Y\sim N(\mu,\sigma^2)$. 因此

$$
\begin{aligned}
&EX^k=E(e^{kY})\\
&=\int_{-\infty}^{+\infty}e^{ky}\frac{1}{\sqrt{2\pi}\sigma}e^{-(y-\mu)^2/2\sigma^2}dy\\
&=\int_{-\infty}^{+\infty}\frac{1}{\sqrt{2\pi}\sigma}e^{-(y^2-2(\sigma^2k+\mu)y+\mu^2)/2\sigma^2}dy\\
&=e^{k\mu+k^2\sigma^2/2}\int_{-\infty}^{+\infty}e^{-(y-(\sigma^2k+\mu))^2/2\sigma^2}dy,
\end{aligned}
$$

积分部分是 $N(\mu+k\sigma^2,\sigma^2)$ 在实数轴上的积分故为 $1$, 即

$$
EX^k=e^{k\mu+k^2\sigma^2/2}.
$$

代入 $k=1,k=2$ 联立解得

$$
\begin{cases}
\mu=(1/2)\ln\left((\sum_{i=1}^nX_i)^4/(n^3\sum_{i=1}^nX_i^2)\right),\\
\sigma^=\ln\left(n\sum_{i=1}^nX_i^2/(\sum_{i=1}^nX_i)^2\right).
\end{cases}
$$

(6)

由于 $X-\mu\sim E(\lambda)$, 则

$$
EX=1/\lambda+\mu,
$$

$$
\begin{aligned}
EX^2&=DX+(EX)^2\\
&=2/\lambda^2+2\mu/\lambda+\mu^2,
\end{aligned}
$$

令 $EX=\overline X$, $EX^2=\sum_{i=1}^nX_i^2/n$, 则联立解得

$$
\begin{cases}
\mu=\sum_{i=1}^nX_i/n-\sqrt{(\sum_{i=1}^nX_i^2-(\sum_{i=1}^nX_i)^2)/n},\\
\lambda=\sqrt{n/(\sum_{i=1}^nX_i^2-(\sum_{i=1}^nX_i)^2)}.
\end{cases}
$$

## 030205

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本, 求 $P\{X>1\}$ ​的矩估计.

### 030205-解答

已知正态分布的矩估计:

$$
\begin{aligned}
EX&=\frac1n\sum_{i=1}^nX_i=\mu,\\
EX^2&=\frac1n\sum_{i=1}^nX_i^2\\
&=DX-(EX)^2=\sigma^2+\mu^2,
\end{aligned}
$$

解得

$$
\begin{cases}
\widehat{\mu}=\overline{X},\\
\widehat{\sigma}^2=\sum_{i=1}^n(X_i-\overline{X})^2/n,
\end{cases}
$$

则由期望的线性性,

$$
\begin{aligned}
P\{X>1\}&=1-P\{X\le1\}\\
&=1-\Phi\left(\frac{1-\mu}{\sigma}\right)\\
&=1-\Phi\left(\frac{\sqrt{n}(1-\overline{x})}{\sqrt{\sum_{i=1}^n(X_i-\overline{x})^2}}\right).
\end{aligned}
$$

## 030207

设 $X_1,X_2,\cdots,X_n$ 是来自总体 $X$ 的样本, 且 $X$ 的分布函数为

$$
F(x,\theta)=
\begin{cases}
1-x^{-\theta},&x\ge1,\\
0,&x<1,
\end{cases}
$$

其中 $\theta>1$ 为未知参数, 求 $\theta$ ​的矩估计.

### 030207-解答

由 $F(x,\theta)=(1-x^{-\theta})I_{\{x\ge1\}}$, 则

$$
f(x,\theta)=F'(x,\theta)=\theta x^{-(\theta+1)}I_{\{x\ge1\}}.
$$

$$
EX=\int_{1}^{+\infty}\theta x^{-\theta}dx=\frac{\theta}{\theta-1},
$$

令 $EX=\overline{X}$, 则

$$
\widehat{\theta}=\frac{\overline{X}}{\overline{X}-1}.
$$
