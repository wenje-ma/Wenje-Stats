## 040104

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本, 其中 $\mu\in\mathbb{R}$ 为未知参数, $\sigma>0$ 已知, 记

$$
\overline{X}=\frac{1}{n}\sum\limits_{i=1}^n X_i.
$$

则当 $\mu$ 的置信区间为

$$
\left[\overline{X}-\frac{\sigma}{\sqrt{n}}u_{0.05},\overline{X}+\frac{\sigma}{\sqrt{n}}u_{0.05}\right]
$$

时, 求置信水平.

### 040104-解答

$$
\begin{aligned}
&P_\theta\{\overline{X}-\frac{\sigma}{\sqrt{n}}u_{0.05}\le\mu\le\overline{X}+\frac{\sigma}{\sqrt{n}}u_{0.05}\}\\
&=P\{-u_{0.05}\le\frac{\overline X-\mu}{\sigma/\sqrt n}\le u_{0.05}\}\ge1-\alpha,
\end{aligned}
$$

由 $u_{0.05}$ 的定义可知 $1-\alpha=0.9$, 即置信水平为 $0.9$.

## 040105

设 $X_1,X_2,\cdots,X_n$ 是来自正态总体 $N(\mu,\sigma^2)$ 的样本, 其中 $\mu\in\mathbb{R}$, $\sigma>0$ 为未知参数, 记

$$
\overline{X}=\frac{1}{n}\sum\limits_{i=1}^n X_i,\quad S^2=\frac{1}{n-1}\sum\limits_{i=1}^n (X_i-\overline{X})^2,
$$

则当 $\mu$ 的置信上限为

$$
\overline{X}+\frac{S}{\sqrt{n}}t_{0.05}(n-1)
$$

时, 求置信水平.

### 040105-解答

$$
\begin{aligned}
&P\{\mu\le\overline{X}+\frac{S}{\sqrt{n}}t_{0.05}(n-1)\}\\
&=P\{\frac{\overline X-\mu}{S/\sqrt n}\le t_{0.05}(n-1)\}\ge1-\alpha,
\end{aligned}
$$

由 $t_{0.05}(n-1)$ 的定义可知 $1-\alpha=0.95$, 即置信水平为 $0.95$.
